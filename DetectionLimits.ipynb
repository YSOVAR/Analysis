## Look at the cadence of the observations in each cluster in the fast-cadence sampling

We read the observation times from a csv file. This file contain the dates and times of each observation. One column holds the dates in (month-day-year order), a second one the actual time. So, after reading that file, we transform this into Julian Dates and (since we are only interested in the spacing between the observations), subtract the time of the first observation, so that every time scale starts at $t=0$. Below are a few plots that show the cadence of the observations for all clusters in the context of the YSOVAR caompagne. A few of the other, not directly related, but affiliated observations (such as the CSI data) are also shown. Finally, we compare a histogram of the time spacings of the individual clusters for the fast-cadence data.

```python
obstimestab = ascii.read(inroot+'/cadences.csv')

def tab2isot(date, time):
    date = date.split('/')
    return '{0}-{1}-{2} {3}'.format(date[2], date[0], date[1], time)

obstimes={}
for region in set(obstimestab['Target']):
    ind = (obstimestab['Target'] == region)
    obstimes[region] = Time([tab2isot(d,t) for d,t in obstimestab['obs start day', 'obs start time'][ind] ], scale='utc').jd - 2400000.5 
    #obstimes[region] -= np.min(obstimes[region])
```

```python
# function to simulate lightcurves
def lcwhitenoise(t):
    return np.random.normal(size = len(t))

def lctrend(t, a):
    slope = a / (np.max(t)-np.min(t))
    return t * slope

def lcperiod(t, a, period = 1.):
    return a * np.sin(t * 2. * np.pi / period)

def lcstep(t, a, p=0.5):
    return a * (np.random.uniform(size=len(t)) > p)
```

```python
# functions to collect the results for the simulated lightcurves
def analyse_fake_lc(lcfunc, a, t, **kwargs):
    maxfreq = kwargs.pop('maxfreq', 1.)
    lc = lcfunc(t, a, **kwargs)
    lc1 = lc + lcwhitenoise(t)
    lc2 = lc + lcwhitenoise(t)
    err = np.ones_like(lc1)
    chi = redchi2tomean(lc1, err)
    stet = stetson(lc1, lc2, err, err)
    period, peak, FAP = lombscargle(t, lc1, maxfreq=maxfreq)
    p = 0 if FAP > 0.05 else period
    return chi, stet, p

def run_sims(lcfunc, n_sim, amp, t, **kwargs):
    stats = np.zeros((n_sim, len(amp), 3))
    for ia, a in enumerate(amp):
         for j in range(n_sim):
            stats[j,ia,:] = analyse_fake_lc(lcfunc, a, t, **kwargs)
    return stats
```

```python
# broadcast those functions to the engines for parallel execution.
dview['lcwhitenoise'] = lcwhitenoise
dview['lcstep'] = lcstep
dview['lcperiod'] = lcperiod
dview['lctrend'] = lctrend

dview['analyse_fake_lc'] = analyse_fake_lc
dview['run_sims'] = run_sims
```

```python
def plot_images(stat, extent, cmap = plt.get_cmap('binary'), ylabel = '', period=None):
    aspect=(extent[1]-extent[0]) / (extent[3]-extent[2])
    fig = plt.figure(figsize=(15,4))
    ax1 = fig.add_subplot(131)
    ax2 = fig.add_subplot(132)
    ax3 = fig.add_subplot(133)
    ax1.imshow(((stat[:,:,1,:] > .9).sum(axis=0, dtype=np.float)/n_sim).T, aspect=aspect, vmin=0., vmax=1., origin='lower', extent=extent, cmap=cmap)
    ax1.set_title('Stetson > 0.9')
    ax2.imshow(((stat[:,:,0,:] > 5.).sum(axis=0, dtype=np.float)/n_sim).T, aspect=aspect, vmin=0., vmax=1., origin='lower', extent=extent, cmap=cmap)
    ax2.set_title('$\chi^2$ > 5')
    ax3.imshow(((stat[:,:,2,:] > 0).sum(axis=0, dtype=np.float)/n_sim).T, aspect=aspect, vmin=0., vmax=1., origin='lower', extent=extent, cmap=cmap)
    ax3.set_title('LS period found (could be the wrong one)')
    for ax in [ax1, ax2, ax3]:
        ax.set_xlabel('amplitude relative to noise')
        ax.set_ylabel(ylabel)
```

In each panel in the following figures I show the statistic properties for a grid of simulated lightcurves. For each point in the grid I simulate n_sim (e.g. 100 or 1000) two-band lightcurves. I calculate the Stetson index, the $\chi^2$ values and the Lomb-Scargle periodogram for each of them. For each grid-point I calcualte the fraction of lightcurves that are flagged a variable accoring to the YSOVAR criteria. Each figure has a color scale from shite to black. White mean none of the lightcurves simulated for this grid point are flagged as variable, black means all of them are flagged as variable. Shades of gray show fractions between 0 and 1. From left to right the panels show variablity as found with the Stetson index > 0.9, the $\chi^2 > 5$ for a single band lightcurves and some period was found in the LS periodogram with FAP < 0.05. This plot does not show it, but I checked the periods that where found and there is only a negligible fraction of periods that do not recover the period I put into the simulations.

For each grid, the x-axis is the amplitude of the signal (in the case of periodic lightcurves the amplitude of a sin function) in relation to the noise. Looking at the first figure, we wee that a periodic signal where the amplitude of the signal is at least twice as big as the stadard deviation of the noise will be flagged as veriable. Physically, this could be bright star (with small measurement uncertainties) that varies only little, or a faint star (with larger measurement uncertainties), that varies a lot.

\subsection{Detection limits for variability}
In section~\ref{sect:indentifyingthevariables} we describe how variable sources are identified using the Stetson test, or, if no simultaneous data in IRAC1 and IRAC2 is available, using the $\chi^2$ test. The cut-off values for those tests are desinged to yield a conservative list of variable sources and to reject those sources where the variability stems mostly from instrumental artifacts. We perform Monte-Carlo (MC) simulations to quantify how much variablity is required to meet those criteria. 

In real sources several different physical effects may contribute to the observed variability and this can lead to very complex patterns in the lightcurve. In the absence of a theoretical model to explain the different contributions, we concentrate on simple analytical prescriptions for lightcurves to gain a sense of the sensitivity of our statistical tools. First, we consider a sources that has two states, a bright and a faint state. The lightcurve switches randomly between those two states. We simulate lightcurves for a different fraction of time spend in the uppe state (0\%, 10\%, 20\%, 30\%,40\% and 50\%) and we expect that variability is found easier for sources with a larger amplitude between the two states and an equal chance to find the source in each state. Second, we simulate sinosoidal lightcurves with different periods (0.1 to 2 days in steps of 0.1 days and steps of 2 days after that). We add Gaussian noise to each lightcurve and vary the ratio of the amplitudes of the signal and the noise (from 0 to 10 in steps of 0.1, where a relative amplitude of 0 means a constant lightcurve with noise only). The sensitivity of the Stetson test, the $\chi^2$ test and the Lomb-Scargle periodogram is independent of the absolute magnitude of a source, only the relative amplitude of the signal and the noise plays a role. Thus, it is equally likely to detect strong variability in a weak source (with large photometric uncertainties) as weak variability in a bright source (with small photometric uncertaintes).
The simulations are done for a grid of relative amplitudes and fractions of time in upper level or period. We simulate 10000 ligthcurves for each grid point for each band. The lightcurves are sampled at the time intervals of the actual YSOVAR observations. Figure~\ref{fig:dtpercluster} shows the histogram of the time steps between observations for each cluster. Clearly, the histograms fall in two groups with AFGL~490, Mon~R2 and NGC~2264 having a lower sampling rate than the other clusters.

The sampling is very similar for most clusters, but AFGL~490, Mon~R2 and NGC~2264 have a lower sampling rate. \label{fig:dtpercluster}}

However, for all clusters the sampling is is non-uniform to avoid aliasing for a specific period. We run all MC simulations using the actual sampling from L1688, one of the clusters with a high sampling rate and Mon~R2, a representative cluster with a low sampling rate. 

\begin{figure*}
\plotone{detectionlimits_L1688}
\caption{Effiency for variability detection. The left panel shows simulations for lightcurves with two distinct states (bright and faint), the other two panels show results for a sinusoidal lightcurve.
The time sampling matches te observations of L1688.
The x-axis shows the ratio of the signal amplitude and simulated Gaussian noise. 
\emph{Left and middle:} Colors show the fraction of lightcurves where variability was detected using the Stetson test. Note that the color scale is very non-linear. 
Black contours show the fraction of variability detected with the $\chi^$ test. 
\emph{Right:} The color shows the fraction of lightcurves where the simulated period was recovered in a Lomb-Scargle periodogram. \label{fig:detectionlimits_L1688}}
\end{figure*}
Figure~\ref{fig:detectionlimits_L1688} shows results from the MC simulations using the sampling of the L1688 cluster. The left panel presents the detection efficiency for lightcurves from an object with two distinct luminosities. If the star is found in each state half the time, then the Stetson test will identify it as variable in almost all cases (99\%) if the step size is at least three times larger than the noise level. Since the $\chi^2$ test uses data from one lightcurve only, the step size must be larger (five times the noise level) to reach the same detection efficiency. If the star spends less than 20\% of the time in either state, then there is a resonable chance that the variability will not be found, even for larger step sizes, since the sampling might catch only few datapoints in this state. For a periodic lightcurve (middle panel) variability is again found easier in the Stetson test than in the $\chi^2$ test and the period of the variability does not influence the detection efficency since those two tests do not consider the time ordering of the observed data. Using a Lomb-Scargle periodogram we are sensitive to periods $p$ between about 1 and 15 days and even periods where the amplitude $a$ of the lightcurve $a \sin(\frac{2\pi}{p}t)$ is only twice as large as the noise level are easily detected, almost independent of the period in the range 1 to 15 days. Such weak signals would not necessarily show up as variable in the Stetson or $\chi^2$ test (see middle panel). In general, the region where the tests detect variability in some lightcurves but not in others is fairly narrow. 

\begin{figure*}
\plotone{detectionlimits_MonR2}
\caption{Same as in figure~\ref{fig:detectionlimits_L1688}, but using the time sampling of the Mon~R2 observations. \label{fig:detectionlimit_MonR2}}
\end{figure*}
Figure~\ref{fig:detectionlimits_MonR2} shows the same plots as figure~\ref{fig:detectionlimits_L1688} but for the time sampling of Mon~R2, one of the clusters with a lower cadence in the observations. The general shape of the regions in the parameters space where variability can be detected is the same, but due to the lower number of observations and the larger time span between observations a larger amplitude is required and we are not sensitive to periods below about 2 days.

To explore more complicated lightcurves we combined several effects, e.g.\ a sinusiodal periodicity overlayed on a long-term trend. In these cases, generally the strongest effect determines how the variability will be seen. If the magnitude of the trend is large and the amplitude of the sinus is small, then the lightcurve will be marked as variable, but the periodicity might not be detected. Given the possible complexity of real lightcurves, it is not possible to cover the entire parameter space with MC simulations. However, the scenarios presented here show that in general we can detect periodicity with an amplitude just twice the level of the noise, variability at three to five times the level of the noise with the Stetson test and about twice as large for single-band lightcurves with the $\chi^2$ test. A similar ratio in the sensitivity of the Stetson and the $\chi^2$ test was also found by \citet{2013AJ....145...66F}, although they used different cut-off levels than this work.

MC11 already found that there is an almost one-to-one relation between the mean magnitude of a source and its mean error in the YSOVAR observations of Orion. Figure~\ref{mag_uncertainty_L1688} shows this relation using all observed sources in L1688. Most sources a concentrated at the lower end of the distribution, only few sources have much larger uncertainties because they are close to bright neighbors, chip adges or effected by other instrumental artifacts. There are only slight variations between clusters which are caused by different levels of diffuse emission and a different severety of instrumental artifact from e.g.\ residual pull-down. For sources brighter than 13~mag the total uncertainty is dominated by the error floor introduced in section \ref{sect:errors}. For those sources we can detect periodicity if the amplitude is larger than about 0.02~mag and variability if it is larger than about 0.03-0.1~mag (with the exact number depending on the signal shape).
