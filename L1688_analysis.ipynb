```python
def planck(T, wave):
    return 2.*const.h*const.c**2. / wave**5. / (np.exp(const.h*const.c / (wave*const.k_B*T)).decompose() - 1.)
```

```python
# Fraction of variable sources
from scipy.optimize import brentq
    
def err_bar(k, N, significance = 0.68):
    '''Estimate uncertainty on probability p=k/N, given k and N

    Given an observed number of sucesses k in N trials,
    what is the probability p for sucess?
    Obviously, p=k/N is a good guess. So, what is the
    uncertainty of that guess?
    For a known p, the probability for observe k=k(N,p) is a 
    binomal distribution. To turn this question around, we have to ask:
    How likely is a binomal distribution with probability p to produce
    the observed outcome k,N? 
    This function numerically approximates that
    distribution (called ``fx`` in the function).
    While it is useful to look at the full distribution of p's,
    that is hard to show in plot. To distill it into a simple number
    this function determines the boundaries p0,p1, such that
    fx(p0) = fx(p1) and $\int_p0^p1 fx(p) dp = significance$.

    Parameters
    ----------
    k : int
        number of sucesses
    N : int
        number of trials
    significance : float
        significance interval

    Returns
    -------
    p0, p1 : float
        boundaries of probability p, such that
        fx(p0) = fx(p1) and $\int_p0^p1 fx(p) dp = 0.68$.
    '''
    if not ((significance >=0.) and (significance<=1.)):
        raise ValueError('Significance must be in interval 0..1')
    x = np.linspace(0,1,1000)
    fx = x**k * (1.-x)**(N-k)
    fx = fx/fx.sum()
    pcrit = brentq(lambda y: (fx[fx >y]).sum()-significance, 0., np.max(fx))
    p0 = np.min(x[fx > pcrit])
    p1 = np.max(x[fx > pcrit])
    return np.array([p0,p1])-float(k)/float(N)
```

For some sources the CLUSTERGRINDER pipeline does not extract photometry in one or more bands. However, extraction of that data might be possible with different, although possibly less precise settings. We use the data released by the c2d survey for this purpose. We only keep data with the data quality flags A, B or C and fill those bands that are missing in the data reduction by Gutermuth et al.
To make this obvious, I implement a simple loop here.

```python
c2d = astropy.io.ascii.read(inroot+'l1688_c2d.tbl', Reader=astropy.io.ascii.Ipac)
magslist = YSOVAR.atlas.sed_bands
zeroflux = {'ir1': magslist['3.6mag'][2], 'ir2': magslist['4.5mag'][2], 'ir3': magslist['5.8mag'][2], 'ir4': magslist['8.0mag'][2], 'mp1': magslist['24mag'][2]}
for n in zeroflux:
    c2d[n+'_d_flux_c'] = -2.5* np.log10(1.-c2d[n+'_d_flux_c']/c2d[n+'_flux_c'])
    #Factor 1e-3 because c2d is in mJy
    c2d[n+'_flux_c'] = -2.5* np.log10(c2d[n+'_flux_c']*1e-3/zeroflux[n])
c2d.remove_columns(['region', 'alpha','j_flux_c',  'j_d_flux_c', 'j_q_det_c', 'h_flux_c', 'h_d_flux_c', 'h_q_det_c', 'ks_flux_c', 'ks_d_flux_c', 'ks_q_det_c', 'dist', 'angle'])
```

### Time sampling

#### Time sampling for typical sources

(From on of Peter's emails):

If you dig into one of Maria's papers or the original YSOVAR proposal, we chose a repeating 3.5 day cadence with time-steps of 2,4,6,8,10,12,14,16,12 hours with +/- 1 or 2 hour tolerances.

The +/-1 or 2 hour tolerances introduced a very useful randomization of the cadence.

The time-steps were chosen to sample different frequencies evenly (ie, d delta_t /dt = constant).  The smaller time-steps made us sensitive to variability as short as 2 hours (high frequencies; and in the case of Orion it was 4 hours, since it took 4 hours to do a single map), without increasing our overall observing program to many more hours than it was actually executed under, or had much smaller time-baselines.  Similarly the longer time-steps made us sensitive to low frequency variations.

The two items above served a very important purpose - avoid 1-day aliasing.  We were in space and there's no reason to have any cadence aliasing.

The 3.5 day repeating pattern was chosen to fit two cycles every week, which is how Spitzer did it's scheduling at the time - in one week chunks.

The plot above compares the measures of the auto-correlation function (ACF) for IRAC1 and IRAC2. If both bands are correlated (some objects might have strong color variations, but that is the exception), then the ACF for both bands should be similar.

The plot compares the coherence time (the width of the first peak in the re-sampled ACF) and the cumulative sum (a measure with a similar idea, but defined for the non-resampled ACF) for IRAC 1 and IRAC 2. For short times (t < 2. days) it's not quite clear that they are related, but for longer times the number obtained from both channels seem to agree to some extend, indicating that the values measured are real and not just noise.

Another way to plot this is to compare the coherence time and the cumulative sum for a single channel and then see if the statistical distributions are similar in both channels.

## Introduction

Stars form in dense and cool molecular clouds. When the density is high enough in a region, the matter can gravitationally collapse and form a young stellar object (YSO). In the early phases, this infall happens almost radially, and the thick envelope dominates the emission from the object and hides the details of what is going on within (class~I). Our knowledge on star formation has developed tremendously in the last few years, not the least due to new observational opportunities on ground- and space-based infrared observatories. In the IR the dust opacity is much less severe than in the optical and thus observations at longer wavelengths are particularly suited to observe YSOs, many of which were discovered in \emph{Spitzer} surveys. Due to the conservation of angular momentum the envelope eventually flattens out to from a circum-stellar disk. If the inclination of the system is not edge-on, then the photospheric emission from the central YSO can be observed. However, the disk still causes an IR excess above the level of a stellar photosphere (class~II), which can be used to distinguish those objects from more evolved stars, e.g. using the \emph{Spitzer} InfraRed Array Camera \citep[IRAC][]{2004ApJS..154...10F} camera. The accretion disk does not reach down to the central star. Instead, its inner radius is given by the dust-sublimation temperature. Some of the mass in the circum-stellar disk is accreted onto the central star via magnetically confined accretion funnels, some is blown out by disk winds and some condenses into planets. When the disk is cleared, the IR colors match those of main-sequence stars (class~III).

In addition to the circum-stellar absorption many YSOs are embedded in the molecular cloud, so that even class~III objects often appear reddened and are invisible in the optical. The closest optical counterpart of the class~II sources are T~Tauri stars (TTS), originally identified by their variability \citep{1945ApJ...102..168J} -- long before anybody realized that TTS are indeed pre-main sequence stars. In most cases, particularly in class~II and III sources, the dominant time scale in the optical is the stellar rotation period, typically a few days \citep{1983ApJ...267..191R,1986A&A...165..110B,2009ApJ...694L.153N}. YSOs can have both cool spots caused by magnetic activity similar to our Sun and hot spots which mark the impact points of the accretion funnels onto the stellar surface. This impact happens at free-fall velocities up to 500~km~s$^{-1}$, thus the accretion shock heats the accreted mass to X-ray emitting temperatures \citep[e.g. review by][]{2011AN....332..448G}. In the optical, the accretion region appears as emission that can be approximated as a black-body with $10-20\,000$~K (\citealp{calvetgullbring}, but see also \citealp{2012AstL...38..649D,2013AstL...39..389D}). Variability in the mass accretion rate can lead to changes in the hot spot signatures. We expect the rotation period of the inner disk where the accretion funnels start to be the typical time scale for those processes and indeed optical variability with amplitudes around 0.1~mag is often observed in CTTS on this time scale \citep{1996MNRAS.282..167S,2008MNRAS.391.1913R}. Another source of variability related to the accretion could be shock oscillations. This has been predicted theoretically \citep[e.g.][]{2008MNRAS.388..357K}, but is not observed so far \citep{2010A&A...518A..54G}. However, \citet{2011AJ....142..141B} find indications that strong accretion in V1647~Ori could excite radial pulsations of the star itself.

YSOs can also vary on much longer time scales. Variability on the scales of years could be caused by changing circumstellar extinction \citep{2007A&A...461..183G} or by massive accretion events when a significant fraction of the disk mass drains onto the YSO. In this case the accretion luminosity can outshine the YSO by orders of magnitudes and it takes months to years \citep[in the case of EXor outbursts,][]{2012ApJ...749..188L} or even centuries \citep[FUor outbursts,][]{1996ARA&A..34..207H} until the accretion decays back down to the original level.

The spectral energy distribution (SED) of YSOs in the optical is dominated by the stellar photosphere and the accretion spot. Thus, optical monitoring is very effective to understand the stellar rotation and the accreting spot. The disk radiates mostly at longer wavelengths. Depending on the mass in the disk and the size of the inner hole it will start to dominate the SED at $K$ band or in the IRAC bands at $3.6\mu$m and $4.5\mu$m. Simple disk models still treat the disk as a static and axisymmetric structure, but observationally it now seems that the disk is in fact ``a bubbling, boiling, wrinkled, dented, warped mass of gas and dust'' \citep{2013AJ....145...66F}.

The YSOVAR (Young Stellar Object VARiability) project has monitored the \object{Orion nebular cluster} and eleven smaller star forming regions with \emph{Spitzer} in the warm mission in $3.6\mu$m and $4.5\mu$m to understand this variability. Data from several different observing programs are analyzed in the context of the YSOVAR project. First results on the \object{Orion Nebular cluster} (ONC) are published in \citet{2011ApJ...733...50M} and \citet{2012ApJ...753..149M}. More details of the observing strategy and an overview of the data can be found in \citet{luisa} (from now on ``the YSOVAR paper''). In this article, we present global properties of the variability in \object{L1688}. Notable individual objects are studied separately. 

Lynds~1688 (L1688) is a sub-cloud of the $\rho$~Oph star forming region, one of the best-studied young clusters in the sky \citep[review by][]{2008hsf2.book..351W}. The distance to the cloud is a matter of active discussion. VLBA measurements and Hipparcos parallaxes point to values around 120~pc \citep{2008ApJ...675L..29L,2008A&A...480..785L}, while \citet{2008AN....329...10M} find 135~pc, again based on Hipparcos data. One possible solution is that the streamers of L1688 are closer to the Sun than the core of the region \citep{2008A&A...480..785L}. In this article, we adopt a distance of 120~pc, but we note that most of our results are independent of the actual distance. The central region of L1688 is very dense and deeply embedded ($A_V=50-100$~mag, see figure~\ref{fig:map}), thus no complete survey of all cluster members exists, but an extinction limited, spectroscopic survey \citep{2011AJ....142..140E} finds an average age of 3.1~Myr for a 6.8~pc$^2$ region centered on L1688 and no significant deviation from the interstellar mass function. Earlier studies \citep{1995ApJ...450..233G,1999ApJ...525..440L, 2002A&A...393..597N} concentrated on the deeper embedded core and found a much younger age of 0.3~Myr. Some, but not all of this difference is due to the specific reddening laws or pre-main sequence evolutionary tracks used in these studies \citep[see discussion in][]{2011AJ....142..140E}.

\citet{2012PASP..124.1137F} present near-simultaneous spectroscopic monitoring for five YSOs in L1688. They do not see any correlation between the hydrogen emission lines that are accretion indicators and the features in the IR lightcurves of their targets indicating that the relatively modest variability in the observation period is not caused by changes in the accretion rate. Notable objects with well-sampled NIR lightcurves are \object{WL 4} \citep{2008ApJ...684L..37P} and \object{YLW 16A} \citep{2013A&A...554A.110P} which show eclipses with periods of 131 and 93~days, respectively, that can be interpreted as signatures of the status as multiple systems, where one or more components are eclipsed by a warped circumstellar disk.

Three fields in L1688 were observed with \emph{Spitzer} in four observing periods from 2010-04-12 to 2010-05-16, 2010-09-22 to 2010-10-27, 2011-04-20 to 2011-05-23 and 2011-10-01 to 2011-11-06. These periods are consecutive visibility periods dictated by the \emph{Spitzer} orbit \citep{2004ApJS..154....1W}. In the first observing period the sampling is much denser. About 70 epochs have been observed on most sources with irregular time intervals to aid period detection and avoid aliasing with the observation frequency in contrast to ground-based data that is frequently effected by an alias due to the day-night cycle on the earth. In the first window, we used a repeating pattern every 3.5 days for easy 1-week scheduling.  Within the 3.5 day window the time-step increased by two hours starting with 2 hours.  In the last three visible windows, the time steps were linearly increasing of 1, 2, 3 days etc.\ to again sample multiple variability frequencies equally. Less than ten datapoints per source were taken in each of the later three observation periods. In total there are 108 epochs of observation with total exposure time of 30.7 hours. All epochs can be found under Program Identification umber (PID) 61024 in the Spitzer Heritage Archive. The three fields chosen were observed with the IRAC1 and IRAC2 channels (effective wavelengths 3.6$\mu$m and 4.5$\mu$m). Both channels operate simultaneously, but their fields-of-view are non-overlapping. Thus, each field is observed in two consecutive pointings, one for each channel. In each observation of one of the target fieldss we automatically obtained data from a neighboring field observed by the channel that is offset from the target field. In the Aprli to May monitoring blocks sources south of the main fields only have IRAC1 data, while those to the north only have IRAC2 data. In the September to November monitoring blocks, the situation is reversed. In each epoch data was taken in IRAC mapping mode using the high-dynamic range (HDR) data acquisition, which obtains an 0.4~s and an 10.4~s exposure for each pointing.

A very detailed account of the data processing and the source extraction for the data is given in the YSOVAR paper and we refer to that publication for details. Here we summarize the main data reduction and processing steps. Basic calibrated data (BCD) are obtained from the \emph{Spitzer} archive. Further data reduction is performed with the IDL package cluster grinder \citep{2009ApJS..184...18G}, that treats each BCD image for bright source artifacts, and combines images into mosaics, rejecting cosmic rays in the process. For each epoch and each channel, aperture photometry is run in individual BCDs with an aperture radius of $2\farcs4$. The photometry reported for an epoch is the average brightness of all BCDs within that epoch that contain the source in question, after rejecting outliers (e.g.\ due to cosmic rays). The photometric uncertainties obtained from the aperture photometry are, particularly for faint sources, only lower limits to the total uncertainty, since distributed nebulosity often found in star forming regions can contribute to the noise. To improve these estimates the YSOVAR paper introduces an error floor value that is added in quadrature to the uncertainties of individual photometric points. The value of the error floor is 0.01~mag for IRAC1 and 0.007~mag for IRAC2.

We cross-match sources from individual observations with a matching radius of $1\arcsec$ with each other and with the 2MASS catalog, which is used as a coordinate reference. All photometric measurements performed in the context of the YSOVAR project are collected in a central database, which we intend to deliver to the IRSA for general distribution. 
Data for this article was retrieved from the YSOVAR database on 22-Aug-2013 and further processed using custom routines in python. ### Instrumental effects in IRAC lightcurves

```python
def characterize_seasonal_jump(t, m, merror):
    '''Characterize the seasonal jump (an instrumental effect)

    This is +1, is the mean in season 1 and 3 is higher than the mean in season 2 and 4,
    -1 if it is vice-versa and 0 if the mean values for each season do not correlate with the 
    time of year the obervations were taken.
    ``np.nan`` is returned for lightcurves that do not have data in all seasons.
    '''
    seasons = np.array([[55295,55338], [55458,55500], [55668, 55710], [55830, 55890]])
    mean = np.zeros(4)
    n = np.zeros(4)
    for i in range(4):
        ind = (t > seasons[i,0]) & (t < seasons[i,1])
        mean[i] = np.mean(m[ind])
    ind = np.argsort(mean)
    if (set(ind[0:2]) == set([0,2])) & (np.min(mean) > 0):   # the mean > 0 makes sure that data exisits in all seasons.
        return 1
    elif (set(ind[0:2]) == set([1,3])) & (np.min(mean) > 0): 
        return -1
    elif (np.min(mean) > 0):
        return 0
    else:
        return np.nan
    
YSOVAR.registry.register(characterize_seasonal_jump, error = True, time=True, force=True, name='seasonaljump')
```

Despite the careful data reduction described above, some residual artifacts remain in the \emph{Spitzer} lightcurves. Compared with other clusters in the YSOVAR project L1688 is particularly suited to discover those since most instrumental artifacts are closely related to the position on the detector. The three target fields observed in L1688 have almost no rotation within one season so that each source falls onto the same region on the chip. The spacecraft orientation flips between seasons, so that most instrumental artifacts produce lightcurves that have one baseline level in season one and three and another in season two and four. Figure~\ref{fig:instrumentallcs} shows examples of lightcurves that might fall into this category.
\begin{figure}
\plotone{instrumentallcs}
\caption{Examples of lightcurves with apparent jumps between seasons. The examples shown here are all $4.5\mu$ lightcurves, but the same can be seen in the $3.6\mu$ channel for
some sources.\label{fig:instrumentallcs}}
\end{figure}
Sufficiently strong intrinsic variability can mask this effect and if the source is intrinsically variable with a period of one year it can mimic this effect. Thus, we visually inspected every frame for a sample of sources with lightcurves similar to those in the figure. Different sources are influenced by different instrumental effects. Some are close to a detector edge, so that the background is not well-determined, some are in the wings of the point-spread function (PSF) of a bright neighbor and some show residuals from hot pixels. There are other sources which appear perfectly normal when looking at the frames. In those sources the variability could either be real or they might fall in a region where the large scale gain-map correction is not accurately calibrated. Most of these effects would only affect one of the spacecraft orientations; even the PSF wings change with the orientation since the PSF is not perfectly circular. Figure~\ref{fig:instrumentallcs} shows sources with a magnitude between 8 and 15 in $4.5\mu$. The relative difference in magnitude is smaller for brighter sources, but since the photometric errors also are smaller these instrumental effects can still be significant. 

In order to quantify the number of lightcurves that suffer from this problem we use a statistical approach. For each channel we have about 200 lightcurves with datapoints in all four observing seasons. If the observed magnitude of a source depends on its position on the detector, then it will be bright in season 1 and 3 and faint in in season 2 and 4 (or faint in 1 and 3 and bright in 2 and 4). We calculate the mean magnitude in each season and test weather a sources follows either of these patterns. If the difference in magnitude between seasons is only observational noise, we expect that $\frac{1}{6}$ of all lightcurves are brighther in season and and three than in season two and four, $\frac{1}{6}$ are brighter in season two and four than in season one and three and $\frac{2}{3}$ of all sources follow a different pattern (the two brightest seasons could be one and two, one and four, two and three, or three and four). Given the large number of sources with different intrinsic time scales even sources with intrinsic variability should have the distribution of $\frac{1}{6}:\frac{1}{6}:\frac{2}{3}$. In contrast, the observed lightcurves have the ratio 49:70:79 for IRAC1 and 57:68:103 for IRAC2. This is incompatible with the expected multinomial distribution (p-values $<10^{-10}$). Comparing the expected and the actual distributions, we see that in each band 50 lightcurves, a quarter of the sample, are missing from the last bin. This implies that about a quarter of all lightcurves suffers from the artifacts described above. It is almost equally likely to find the maximum brightness in season one and three or two and four. However, most of the artifacts depend on the specific detector and thus the probability to find correlated variability in both channels is only about $0.25^2/2=3$\%. 

In section~\ref{sect:variability} we determine the variable members using a Stetson and a $\chi^2$ test as defined in the YSOVAR paper. The limits in those tests are designed to be conservative and indeed we find a distribution of 6:11:27 for IRAC1 and 5:8:32 for IRAC 2, when we restrict the sample to those lightcurves that will be classified as variable below. Both ratios are fully compatible with the expected multinomial distribution $\frac{1}{6}:\frac{1}{6}:\frac{2}{3}$ with a p-value of 0.3 and 0.7, respectively. This shows that the limits we apply are conservative enough that the sample of variable stars only contains sources where the variability is dominated by the intrinsic source variability.

The other star forming regions observed by the YSOVAR project have a higher field rotation within one season, so there is no discrete switching point and the instrumental artifacts discussed here will not result in a step in the lightcurve. This makes them more degenerate with intrinsic stellar variability, but the observations all use the same setup and the same definition for intrinsic variability as we do here, so we expect that the sources classified as variable in other YSOVAR studies also have a negligible contribution from instrumental artifacts.

Only 77 of all 984 sources in this study are members of the L1688 star forming regions according to \citet{2008hsf2.book..351W}. Just over 10\% of the 984 sources are classified as stars by \citet{2009ApJS..184...18G}. Of those sources, 17 are class~I sources, 36 are class~II sources and 59 are class~III, i.e.\ disk-less sources, which can either be pre-main sequence stars that lost their disks only recently, or main-sequence stars. The \citet{2009ApJS..184...18G} classification scheme is conservative, it represents a set of sources that can be dereddened and classified to a high degree of confidence. However, for many sources we lack the required spectral coverage. Thus, the YSOVAR paper defines a simpler approach, which uses the observed colors only. In this scheme, we fit the spectral slope $\alpha=\frac{d \log \lambda F_{\lambda}}{d \log \lambda}$, where $\lambda$ is the wavelength and $F_{\lambda}$ the flux per unit wavelength interval at that wavelength. We make use of all measured fluxes (no upper limits) in the 2-24$\mu$m range, which corresponds to the range from the $K$ filter to the MIPS 24$\mu$m channel. We use the fluxes for all \emph{Spitzer} bands from \citet{2009ApJS..184...18G} or c2dr and summarize the new lightcurves by calculating the mean and standard deviation for the new IRAC1 and IRAC2 data. Thus, sources detected both in the cold mission and in the new dataset, will have two datapoints for IRAC1 and IRAC2. We perform a least-squares fit and call sources with $0.3 < \alpha$ class~I, $-0.3 < \alpha <0.3$ flat-spectrum (F), $-1.6<\alpha < -0.3$ class~II and $\alpha < -1.6$ class~III sources. In this scheme a significantly reddened main-sequence star, which has an intrinsic slope $\alpha<-1.6$, may appear as class~II, flat-spectrum or even class~I object, so this observational classification cannot be translated directly into the evolutionary stage of an object. Detections in any two bands are sufficient to calculate $\alpha$, thus we can assign a classification to 895 of the 984 objects. We find 108 class~I, 58 flat-spectrum, 285 class~II and 444 class~III sources. Comparing those sources, which are also identified by \citet{2009ApJS..184...18G}, the resulting classification is very similar, particularly for the class~I and flat-spectrum sources. All sources with a spectral slope $<0.3$ according to our slope-fitting are class~I sources in both classification schemes. With two exceptions our flat-spectrum sources are either class~I or II in \citet{2009ApJS..184...18G}. Of the 43 sources with a spectral slope between -1.6 and -0.3, 24 are also called class~II by \citet{2009ApJS..184...18G}, but 18 are class~III in that paper, indicating that these may be reddened background stars. The general agreement between the more complex classification scheme and the observed spectral slope is also found for other star forming regions (YSOVAR paper).

We use three different methods to detect variability in the lightcurves. For details on those methods see the YSOVAR paper. Sources in the target fields are observed in IRAC1 and IRAC2 within a few minutes. For those sources we calculate the Stetson index $s$ with points weighted evenly \citep{1996PASP..108..851S}.
\begin{equation}
s = \frac{1}{\sqrt{N(N-1)}}\sum_{k=1}^{N}sig(P_k)\sqrt{|P_k|}
\label{eqn:stetson}
\end{equation}
where the sum is taken of all $N$ pairs of observations in IRAC1 and IRAC2 with observed magnitudes $a_k$ and $b_k$ and uncertainties $\sigma_{a_k}$ and $\sigma_{b_k}$. $sig$ denotes the sign of $P_k$. $P_k$ is the product of the normalized residuals in both bands:
\begin{equation}
P_k  = (\frac{a_k-\bar{a}}{\sigma_{a_k}}) \times (\frac{b_k-\bar{b}}{\sigma_{b_k}})
\end{equation}
Here $\bar{a}$ denotes the error weighted mean of all IRAC1 magnitudes and $\bar{b}$ the error-weighted mean of all IRAC2 magnitudes. The Stetson index is very robust to observational errors since those are unlikely to effect both bands in the same way. In the first season we find 40 sources to be variable, that is 18\% of all lightcurves where we can calculate the Stetson index. The number is slightly higher (47 sources, 20\%) if we consider all observing seasons. This is not surprising, given that the Stetson index is designed to be a ``robust'' statistic that is little influenced by a small number of datapoints and  most datapoints belong to the first season. We caution that equation~\ref{eqn:stetson} will introduce a systematic bias in such a way that a source that is variable in only one band will not be marked as variable. 
Figure \ref{fig:stetsonlcs} shows two examples of lightcurves that are classified as variable according to the Stetson index.

For sources outside the primary target fields or for sources which are so bright that one IRAC channel is saturated the Stetson index cannot be calculated, since lightcurves are observed in one only band for any given observing season. For those sources we calculate the reduced $\chi^2_{red}$ to the mean of the lightcurve. Instrumental uncertainties lead to a non-Gaussian error distribution and we use a conservative cut-off and mark sources as variable only if $\chi^2_{red} > 5$ (YSOVAR paper). Using this metric, we find 49 sources that exhibit variability in IRAC1 and 38 that exhibit variability in IRAC2. Examples are shown in figure~\ref{fig:chi2lc}. If instead of comparing to the mean (i.e.\ a constant) we fit a slope, only one source is classified differently. 
When fitting higher order polynomials, the number of sources with $\chi^2_{red} > 5$ goes down, e.g. for a quadratic polynomial only 41 and 30 sources have a $\chi^2_{red}> 5 $ in IRAC1 and IRAC2, respectively. 
Many sources have significantly larger values of $\chi^2_{red}$ for the whole observation series than within one season indicating a constant luminosity over 40~days, but change it over time scales between six months and two years.

Last, we search for periodicity in the lightcurves using a Lomb-Scargle periodogram \citep{1976Ap&SS..39..447L,1982ApJ...263..835S} and the algorithm of \citet{1989ApJ...338..277P}. We require a false alarm probability $<0.05$ (see YSOVAR paper) and only search for periods between 1 and 15 days, so that about three periods fit in each observing season. Comparing IRAC1 and IRAC2, we adopt the period that is found with the higher significance. Allowing longer periods leads to the detection of many long-term trends, where the data does not show that these trends are actually periodic. We find seven sources that show a periodic behavior in the first observing season, but in all cases this periodicity is not stable over more than one season. Five of those sources have information in both IRAC bands and are already classified as variable by the Stetson index, one is variable according to the $\chi^2_{red}$ test and the remaining one is observed in one band only, but this is a faint source, so the $\chi^2_{red} = 2.7$ only. The number of datapoints in the other seasons is too low to search for periodicity within an particular season except the first. Figure~\ref{fig:periodlc} shows the phase-folded lightcurve in one exemplary band for all six periodic sources where the periodicity is significant in a single band. For one source, the period is seen with a low significance in [3.6], but it is more significant in the color term [3.6]-[4.5]. The folded single band and color lightcurves for this source are shown in figure~\ref{fig:periodcolorlc}. Six of the periods are less than 7 days, the remaining one is 15 days. The largest amplitude is around 0.3~mag, the smallest around 0.05~mag. In both figures, the datapoints taken in the first season follow the folded period much better than the remaining data, indicating that the period is not stable for longer than a few months. In source 534, the datapoints of the later seasons, folded with the same period, seem to follow a different, yet clearly defined lightcurve with a similar period. However, due to the low number of points, this is statistically less significant.

Figure~\ref{fig:probvarSED} shows the fraction of variable stars with a mean magnitude $<14$ for cluster members and non-members. For fainter stars the variability cannot be reliably detected, unless the amplitude is exceptionally large. The observed fraction of variable stars is the best-guess for the probability to find variability for a star of this evolutionary class. However, some bins in the histogram contain only a small number of sources, so this probability is not very well constrained. We calculate the uncertainty for the probability based on the observed number of variable stars $v$ in a bin with $N$ total sources as follows: If the true probability for a source in the bin to be variable is $p$, then the probability to observe $v$ variables in $N$ stars is given by $p_{obs}$, which follows a binomial distribution:
\begin{equation}
p_{obs}(v,N,p) = {N \choose v} p^v (1-p)^{N-v}
\end{equation}
The peak of this distribution is $p_{obs}(p) = \frac{v}{N}$. We then calculate the boundaries of the confidence interval $p_1 .. p_2$ for $p$ such that $p_{obs}(p_1) = p_{obs}(p_2)$ and the area under the curve includes 68\% of the probability. Those uncertainty ranges are shown in figure~\ref{fig:probvarSED}. The observed probability for all classes of non-members is compatible with a variability fraction just under 10\%. The fact that this barely changes with the evolutionary class, much unlike the distribution for the cluster members, indicates that the contamination with unidentified cluster members is low since they would bias the observed variability fraction to higher values for the younger classes and the vastly different number of sources in the different bins make it highly unlikely, that each bin is contaminated by the same fraction of class~I-II cluster members.

All known cluster members in class~I are variable. This fraction decreases to about 80\% in class~II sources and then drops rapidly. In class~III sources it is already compatible with the fraction observed for non-members. This distribution is a powerful indication that the main cause for the variability must be found in the inner disk regions. 
In principle the SED classes of YSOs (class~I, flat-spectrum, II and III) represent an evolutionary sequence. Initially, the entire emission from the YSO is dominated by the dusty envelope (class~I). As the system evolves the dust settles in a disk (class~II) until this disk disperses entirely due to planet formation, accretion and evaporation and all IR emission is photospheric. YSOs in class~III cannot be distinguished from main-sequence objects by their IR emission alone. 

In class~I and flat-spectrum sources a significant fraction of the observed IR radiation comes from the envelope, but the variability is still present when the envelope disperses and the mass concentrated in the disk (class~II). However, when the inner disk vanishes, the variability suddenly becomes undetectable. In practice, the observational SED scale that we use here will not follow this path exactly. One complication is that intermediate dust, that is not associated with the specific YSO, also reddens the spectrum and makes the star appear younger than it actually is, e.g. background star that happens to be found behind a this layer of the cloud might appear as class~II or even flat-spectrum source. Also, the orientation of the specific system plays in important role. Already at very early stages (class~I) YSOs can develop powerful outflows, which pierce their envelope and allow us to view the star through the outflow cavity, so that the photospheric contribution becomes important in an early stage. On the other hand, if we see the accretion disk edge-on it will cause a high degree of reddening even for late evolutionary stages, when the envelope is already dispersed. Simulations show that stars of very different ages can thus appears with a similar SED \citep{2006ApJS..167..256R}. These effects tend to smooth out age effects when comparing different evolutionary classes, as each bin will contain a few sources of neighboring evolutionary stages.
Similarly, the membership lists for L1688 might not be complete, so that some sources that we treat as non-members here are actually YSOs.

Figure~\ref{fig:meanampSED} looks in more detail at the variable sources using the $4.5\mu$m data. The plot for $3.6\mu$m looks similar, but is not shown here. The figure shows the median amplitude of the observed variability. The errorbars in this figure represent the standard deviation of the mean amplitudes within one class. 
For members we can clearly see that the variability is stronger in class~I and flat-spectrum sources than in class~II sources, where the amplitude is still larger than in class~III sources. Specifically, the spread of the amplitudes in the earlier classes is much larger than for the later classes. For non-members the observed amplitudes seem to follow a similar, but less pronounced trend. This suggests that this category includes not only less reddenend background objects with an intrinsic SED of class~III, but also contains some number of as yet unidentified cluster members.

In summary we can say that the probability for a source to be observed as a mid-IR variable decreases only a little between class~I and II sources, but the mean amplitude as well as the differences within a class are much larger for younger sources which have more circumstellar material.

As the last aspect in this section we will look at the timing behavior of the lightcurves. The list of periodic sources contains objects of all evolutionary stages, but is too small to recognize any trends in variability time scale with the evolutionary stage. To asses changes in the time-scale in all sorces, not just the periodic ones, we calculate the auto-correlation function $ACF(\tau)$ for each lightcurve. By definition $ACF(\tau=0)=1$ and the $ACF$ decays for longer time-lags $\tau$. We take the first value of $\tau$ with $ACF(\tau)=0.5$ as the characteristic time-scale for a lightcurve. A more common definition would be to use the position of the first local maximum in the $ACF$, but due to the low number of datapoints in our lightcurves, the noise in the auto-correlation function is large and this value is often not very well defined. While our definition might not give us the time scale of the physical processes at work, they are still valid for relative comparisons. One complication here is that our data is unevenly sampled. In order to calculate the $ACF$ we linearly interpolate the lightcurve on a grid with time steps of 0.1~days. This process can change the properties of the lightcurves on short time scales. However, for most lightcurves the relevant time scales are longer than the distance between two observations (0.1 to 0.8 days). We find that the average $ACF$ time scale decreases from 4~days for class I sources to 1.5 days for class~III sources. A more detailed analysis of stochastic and quasi-periodic properties of the lightcurves is beyond the scope of the current paper \citep[see][for a discussion of these properties in $JHK$ lightcurves]{parksetal}.

For sources in the primary target fields the observations in $3.6\mu$m and $4.5~\mu$m are separated by a few minutes only. For those sources we compare the color and the magnitude in the $3.6\mu$m band. Figure~\ref{fig:cmd} shows the color-magnitude diagram (CMD) for four sources. In some cases, the color is fairly stable within one season, but changes between seasons; in others the time scale of variability is much shorter and color changes are seen within one season as well. For non-variable sources the CMD forms a point cloud, but for variable sources the shape of the CMD can reveal the physical cause of the variability, e.g.\ if a disk warp or accretion funnel passes in front of a YSO we expect its color to become redder as it becomes fainter. If the absorber has the same gas and dust properties as ISM, then the datapoints in the CMD would follow a line with the slope of the interstellar reddening law. We fit a straight line to all points in the CMD using an orthogonal distance regression method that takes the errors in both the x and y directions into account. From a total of 47 variable sources with CMDs with a sufficient number of datapoints to perform this fit, 39 can be fitted with a formal uncertainty on the best fit slope below $6^{\circ}$. As can be seen in figure~\ref{fig:cmd} even in those sources with a well defined slope in the CMD the scatter around the best linear fit is not compatible with just the measurement uncertainty: Individual seasons are systematically above or below the line, the spread is larger than the measurement uncertainty and the points do not cluster strogly, instead they fill some regions of the parameter space almost evenly (e.g.\ lower left panel in the figure). Note that the seasonal variation in the upper left panel in figure~\ref{fig:cmd} cannot be an instrumental artifact such as those discussed in section~\ref{sect:instrumentaleffectsiniraclightcurves} because this would cause season two (cyan) and four (red) to be both brighter or both fainter but here season two falls below the fit line and season four above.

## Discussion

### Variability fraction and membership

In section~\ref{sect:variability} we show that most YSOs (class~I to II) in the L1688 star forming region are detected to be variable. The variability fraction drops from 100\% for class~I sources to about 80\% for class~II sources. The expected variability in objects not associated with the star forming regions is only about 5\% as seen in non-member class~III sources. This category is certainly not contaminated by disk-bearing stars, since reddening by the cloud can only increase the spectral slope $\alpha$ but never hide an existing IR excess. The larger fraction of non-member variable objects in the other classes in figure~\ref{fig:probvarSED}, thus shows that the membership lists are incomplete. Indeed, we propose that all variable class~I and, F and II sources are members.

\emph{Spitzer} observations of other young clusters also find that younger stars are more variable, but with lower variability fraction. \citet{2011ApJ...733...50M} find only 70\% of all stars with disks in their ONC sample to have detectable variability; \citet{2013AJ....145...66F} find 60\% in IC~348. Both clusters are located at larger distances and thus it is likely, that they miss low amplitude variability. The membership lists for L1688 are certainly not complete, but there is an obvious bias to preferably detect bright cluster members. Much of the classification in the literature is based on optical spectroscopy, looking for \ion{Li}{1} in absorption or H$\alpha$ in emission and optical spectroscopy is only possible for objects with a low extinction. Due to their high energy, X-rays penetrate significantly larger column densities and can be used to find YSOs down to absorbing column densities up to a few times $10^{22}$~cm$^{-2}$. However, in this specific case the field where the absorption of the cloud is the largest (the field in the north-west), is most distant from the optical axis in the \emph{Chandra} observation and thus \emph{Chandra} is significantly less sensitive in this region. We also classify sources as a member if they show an IR excess. However, even \citet{2009ApJS..184...18G} has a bias to less absorbed sources, because observations in several bands ($3.6\mu$ and $4.5\mu$ or, if one of those is missing, $K$) are required to perform the dereddening procedure that distinguishes between reddened background objects and YSOs. This causes both samples of membership that we use to preferably include objects with low extinction as can be seen in figure~\ref{fig:hist36} and in the distribution of sources in figure~\ref{fig:map}. In L1688 a K5 star with little extinction will have a magnitude of 12.2 in $3.6\mu$ using the evolutionary tracks of \citet{2000A&A...358..593S} and the YSO colors from \citet{2013arXiv1307.2657P}. In this magnitude range variability down to \textbf{put in number here when I have it} (YSOVAR paper) can be detected. In contrast, for the same star in the ONC the variability has to be XXX times larger to be detected. Figure~\ref{fig:meanampSED} shows that there is a wide distribution of amplitudes in YSOs of early classes. Of course, even our study of the close star forming regions L1688 cannot detect variability smaller than XXX at XXX mag, thus we also miss the faint end of the distribution. However, comparing our results with the previous studies it seems likely that essentially all YSOs show substantial mid-IR variability on time scales of days to weeks.

### Long-term variability

YSOs evolve over thousands and millions of years from the initial molecular cloud to main-sequence stars. On scales of hundreds of years, YSOs show FU~Ori outbursts, but this is still longer than the time covered by our observations. For EXOrs with a time scale of a few months the sampling is not very well suited since the individual seasons are too short and the time span between seasons too long to observe the shape of the decay of a EX~Or outburst. The absence of contemporaneous optical photometry does not allow us to uniquely identify our lightcurves with a specific optical phenomenon, but there are examples for a brightening up to one magnitude. Figure~\ref{fig:stetsonlcs} contains a candidate outburst event consisting of a brightness increase by about 0.6~mag and reddening at the same time in SSTYSV J162732.23-242316.0. The rise and fall take only a few days and the entire outburst lasts about a month. 2MASS J16272183-2427275 (= SSTYSV J162721.82-242727.7) presents a slow rise to about 10~mag in $4.5\mu$m and an  apparently smooth, colorless decay over two years (although more activity between seasons cannot be excluded). There is an indication of a slow and smooth increase in brightness again in the last observing season, pointing to a recurring phenomenon. These time scales are much longer then the dynamical time in the inner disk, so they are presumably driven by disk phenomena that originate at larger radii. It is worth noting that AA~Tau, a close-by T~Tauri star, recently showed decline in brightness over scales months or years \citep{2013arXiv1304.1487B} that is well-explained by a warp in the disk at approximately 10~AU. This warp causes extra extinction in the line-of-sight over AA~Tau's almost edge-on disk and reddens the star considerably. In contrast, in the examples discussed here the change is colorless or the star becomes bluer as it is fades, which rules out extinction.

Figure~\ref{fig:cmdslope} shows that the observed slopes in the CMD fall into two categories: Those that become redder as they get fainter and those that become bluer. There is a third category that is not shown in the figure containing those sources that do not have a well-defined slope in the CMD. The reddening group has slopes that are roughly compatible with the slope of an interstellar reddening law. Small deviations can be explained by modifications of the dust-to-gas ratio or the grain size and composition. This is commonly observed in individual young stars \citep{2008A&A...481..735G} and star forming regions as a whole \citep{2010AJ....140..266W,2012AJ....144..101G}. The most prominent example of a class~II source where the extinction changes on time scales of days is \object{AA Tau}. In this case, the absorption dip is periodic and caused by an inner disk warp partially occulting the disk of the star \citep{1999A&A...349..619B,2007A&A...462L..41S}. In the last two years the behavior of AA~Tau changed. It is still variable, but less regular and it appears much fainter. \citet{2013arXiv1304.1487B} explains this with an optical reddening that increased by four magnitudes, most likely due to the some asymmetry in the disk that rotated into the line-of-sight. We also observe additional reddening on time scales of months to years in about half of all sources with as well-defined slope in the CMD. An example of that is shown in the top right panel of figure~\ref{fig:cmd}. We do not know the inclination of the objects in our sample, but if it is close to edge-on, this might be caused by an asymmetric disk similar to AA~Tau. The required warp in a disk or at least a local change in scale height, that lets some gas and dust stick out above the average disk height, can be caused by a planetary mass object orbiting in the disk \citep{2011ApJ...736...85U,2013LNP...861..201B}. Alternatively, vortices in the disk can also push material up locally \citep{2010A&A...513A..60L,2012ApJ...754...21L}. 

Changes in the optical brightness and color of a YSO are often attributed to spots, either hot accretion spots or cold magnetic spots, rotating in and out of view. For two reasons this is not directly relevant in the mid-IR. First, for all YSOs except class~III sources, the mid-IR flux is dominated by emission from the inner disk \citep{2006ApJS..167..256R} and the stellar photosphere contributes only a small fraction, thus changes in the photospheric flux do relatively little to the observed mid-IR magnitudes. Second, mid-IR wavelengths are already in the long-wavelength limit of the Planck function $B(\lambda,T)$ and thus the resulting variations are nearly colorless, in contrast to the observations in figure~\ref{fig:cmdslope}. The exact $\Delta m$ depends on the temperature $T$ of the photosphere, the hot and cold spot as well as the filling factor $f$ \citep{1986ApJ...306..199V,2001AJ....121.3160C}:
\begin{equation}
\Delta m=-2.5\,\log\left(1-f(1-\frac{B(\lambda, T_{spot})}{B(\lambda, T_{*})}) \right)
\end{equation}
As an example we calculate the change im magnitude and color for a stars with $T_{*}=4000$~K.
For a cool spot with $T_{eff} = 3000$~K and a filling factor of $f=10$\% \citep{1989A&A...211...99B,2008A&A...479..827G},  the change in luminosity is $0.04$~mag and the change in color $0.003$~mag. For most stars of our sample fluctuations this small would be masked by the observational noise. The situation is slightly better for hot spots. Assuming a filling factor of 10\% and a hot spot temperature of $20\;000$~K \citep{calvetgullbring}, the change in [3.6] is 0.6~mag, while the [3.6]-[4.5] color changes by 0.05~mag. Again, this is almost colorless and no star in figure~\ref{fig:cmdslope} can be explained by this mechanism alone. If the reddening increases and at the same time the visible fraction of the hot spot decreases, then the resulting slope in the CMD would be intermediate between ISM reddening and colorless. Figure~\ref{fig:cmdslope} shows three class~I source and one class~II source in this region of the CMD slope diagram. If the class~I sources are cluster members and not reddened background objects, then the photospheric contribution in the mid-IR should be negligible. We conclude that spot modulation does not contribute significantly to the mid-IR variability. This can also be one cause for the low number of periodic sources that we observe. Only 8\% of all variable stars in the first observing season are periodic, while \citet{2001AJ....121.3160C} find 18\% in $JHK$ monitoring of the Orion~A molecular cloud with a similar time coverage as we have for L1688 and Parks et al. (in prep.) find a third of all variable stars in L1688 to be periodic (due to their longer time baseline, Parks et al. are sensitive to different periods, but only 4 out of their 32 periods have values that are outside of the range we could detect -- see their table. We are sensitive to periods of up to 15~days here.).

Since spots on the star cannot explain sources that become bluer as they faint, this phenomenon must be related to the disk. Several parameters of the inner disk could be time variable. The most obvious one is the accretion rate. However, the direct effect of increased accretion is a larger accretion spot, which would make the star bluer and brighter, not fainter. Also, \citet{2012PASP..124.1137F} find no relation between mid-IR lightcurves and spectroscopic accretion tracers. This indicates that the parameter that causes the inner disk can change without effecting the accretion rate. In the mid-IR we see the optically thick dust at the inner disk edge. Its temperature is determined by the distance to the central star and its chemical composition. The distance in turn is mostly given by the stellar irradiation and the dust sublimation temperature. Dust that moves inward of the dust sublimation radius is heated and evaporates to gas on time scales below one orbital period \citep[see discussion in][]{2013AJ....145...66F}. If the irradiation is reduced, for example because the accretion rate drops and the accretion spot cools down, then the dust is reformed and the inner disk radius moves inward. Because the disk is heated over its entire vertical height at the inner edge while only the disk surface absorbs radiation at larger radii, it can form a puffed-up inner rim that reaches above the usual disk scale height and casts a shadow on the remaining disk reducing the luminosity at longer wavelengths. If this rim grows, it intercepts more stellar light. Thus the emission from the inner wall increases and the emission from the outer disk decreases, which results in bluer colors. Since the total emitting area also decreases, the source can become bluer and fainter. Time variable self-shadowing has been observed with direct imaging in \object{HD 163296} \citep{2008ApJ...682..548W} and \citet{2010ApJ...717..441E} present a series of \emph{Spitzer} spectra of T~Tauri stars where the spectrum can be modeled taking into account the shadows cast by the inner disk rim. In a time series of spectra those models explain a decreasing in long ($>20\mu$m) flux that coincides with an increase in emission from the inner wall \citep{2011ApJ...728...49E}. A similar mechanism might explain the bluer colors we observe from about half the objects when they become fainter, however, since both $3.6\mu$m and $4.5\mu$m are both formed close to the sublimation radius, this requires changes at those radii, maybe in the shape and not only the height of the inner wall.

Alternatively, \citet{2013AJ....145...66F} suggested a non-axisymmetric model. If the emission from the stellar photosphere is stronger on some longitudes than on others due to a strong accretion spot or magnetic activity at this position, it irradiates the inner disk like a searchlight beam and causes the dust at this position to retreat. The result is a disk where the inner rim is broken at one (or several) positions. This would reduce the emission in $3.6\mu$m, but cause more energy to be emitted at a larger radius, where the dust is cooler on average and thus emits at longer wavelength such as $4.5\mu$m. In this way, the source can become redder as it brightens and bluer when the searchlight beam stops and the dust at the inner rim reforms at all longitudes. In this scenario the time variability of the mid-IR emission should follow the pattern of the searchlight and, if this is emitted by the accretion spot we expect to find the timescale of the stellar rotation period (if the change in brightness is a geometrical effect with the line-of-sight) or a close relation with the accretion rate. Except for the seven periodic sources, we see the lightcurves changing non-periodically with time scales of just a few days, faster than typical rotation periods of YSOs \citep{1989A&A...211...99B,2008A&A...479..827G}. As for the second idea, this would again predict a relation between the mid-IR luminosity and the accretion rate, which has not been observed yet.

None of the simple models presented so far offers a convincing explanation of the observed lightcurves and CMDs. However, detailed simulations that take into account turbulent transport processes in the disk and different dust species with a complete chemical network have not been performed yet. The real situation might be much more complex than sketched above. Since different dust species form and sublimate with different speeds and at different temperatures, time variable irradiation can actually cause a very complex mixture of species with different opacities, that will not react linearly to increased irradiation. Such a complex network might cause a delayed and non-linear response of the disk, which masks the underlying relation between accretion and disk emission.
