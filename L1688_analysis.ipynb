```python
def planck(T, wave):
    return 2.*const.h*const.c**2. / wave**5. / (np.exp(const.h*const.c / (wave*const.k_B*T)).decompose() - 1.)
```

```python
# Fraction of variable sources
from scipy.optimize import brentq
    
def err_bar(k, N, significance = 0.68):
    '''Estimate uncertainty on probability p=k/N, given k and N

    Given an observed number of sucesses k in N trials,
    what is the probability p for sucess?
    Obviously, p=k/N is a good guess. So, what is the
    uncertainty of that guess?
    For a known p, the probability for observe k=k(N,p) is a 
    binomal distribution. To turn this question around, we have to ask:
    How likely is a binomal distribution with probability p to produce
    the observed outcome k,N? 
    This function numerically approximates that
    distribution (called ``fx`` in the function).
    While it is useful to look at the full distribution of p's,
    that is hard to show in plot. To distill it into a simple number
    this function determines the boundaries p0,p1, such that
    fx(p0) = fx(p1) and $\int_p0^p1 fx(p) dp = significance$.

    Parameters
    ----------
    k : int
        number of sucesses
    N : int
        number of trials
    significance : float
        significance interval

    Returns
    -------
    p0, p1 : float
        boundaries of probability p, such that
        fx(p0) = fx(p1) and $\int_p0^p1 fx(p) dp = 0.68$.
    '''
    if not ((significance >=0.) and (significance<=1.)):
        raise ValueError('Significance must be in interval 0..1')
    x = np.linspace(0,1,1000)
    fx = x**k * (1.-x)**(N-k)
    fx = fx/fx.sum()
    pcrit = brentq(lambda y: (fx[fx >y]).sum()-significance, 0., np.max(fx))
    p0 = np.min(x[fx > pcrit])
    p1 = np.max(x[fx > pcrit])
    return np.array([p0,p1])-float(k)/float(N)
```

For some sources the CLUSTERGRINDER pipeline does not extract photometry in one or more bands. However, extraction of that data might be possible with different, although possibly less precise settings. We use the data released by the c2d survey for this purpose. We only keep data with the data quality flags A, B or C and fill those bands that are missing in the data reduction by Gutermuth et al. (2009).

```python
c2d = astropy.io.ascii.read(inroot+'l1688_c2d.tbl', Reader=astropy.io.ascii.Ipac)
magslist = YSOVAR.atlas.sed_bands
zeroflux = {'ir1': magslist['3.6mag'][2], 'ir2': magslist['4.5mag'][2], 'ir3': magslist['5.8mag'][2], 'ir4': magslist['8.0mag'][2], 'mp1': magslist['24mag'][2]}
for n in zeroflux:
    c2d[n+'_d_flux_c'] = -2.5* np.log10(1.-c2d[n+'_d_flux_c']/c2d[n+'_flux_c'])
    #Factor 1e-3 because c2d is in mJy
    c2d[n+'_flux_c'] = -2.5* np.log10(c2d[n+'_flux_c']*1e-3/zeroflux[n])
c2d.remove_columns(['region', 'alpha','j_flux_c',  'j_d_flux_c', 'j_q_det_c', 'h_flux_c', 'h_d_flux_c', 'h_q_det_c', 'ks_flux_c', 'ks_d_flux_c', 'ks_q_det_c', 'dist', 'angle'])

# This code will issue a gazillion warnings like "WARNING: Warning: converting a masked element to nan. [numpy.ma.core]" suppress that outout
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    c2dnames = set(c2d.colnames) - set(('ra','dec'))
    cat.add_catalog_data(c2d, names=c2dnames, ra1='ra', dec1='dec', ra2='ra', dec2= 'dec')

# Rename some of the columns in the c2d catalog to match our (and SIMBAD's) conventions.
irnames = {'ir1': '3.6mag', 'ir2': '4.5mag', 'ir3': '5.8mag', 'ir4': '8.0mag', 'mp1': '24mag'}
for nc2d in irnames:
    channel = irnames[nc2d]
    for i in np.arange(len(cat)):
        if not np.isfinite(cat[channel][i]) and np.isfinite(cat[nc2d+'_flux_c'][i]) and (cat[nc2d+'_q_det_c'][i] in 'ABC'):
            cat[channel][i] = cat[nc2d+'_flux_c'][i]
            cat['e_'+channel][i] = cat[nc2d+'_d_flux_c'][i]   # uncertaintiy
            cat['r_'+channel][i] = 'c2d'                      # reference
    cat.remove_columns([nc2d+'_flux_c', nc2d+'_q_det_c', nc2d+'_d_flux_c'])
cat.remove_columns(['mp2_flux_c', 'mp2_q_det_c', 'mp2_d_flux_c'])
cat.rename_column('object_type','c2d_object_type')
```

```python
# Data was downloaded by Luisa Rebull from
# http://surveys.roe.ac.uk:8080/wsa/region_form.jsp?database=UKIDSSDR9PLUS&programmeID=103&from=source&ra=246.616667&dec=-24.384722&sys=J&radius=1&emailAddress=&rows=30&select=default&where= 
ukidss = astropy.io.ascii.read(inroot+'l1688ukidss.csv')
# Filter out things that are classified as noise, certain galaxy (many artifacts are also classified as galaxy) or as saturated.
# http://surveys.roe.ac.uk/wsa/www/gloss_m.html#udssource_mergedclass
# keep sources classified as noise (the classification is based on colors and YSOs do not have stellar colors, so they might end up in this bin), 
# star, probably star or probable galaxy (because those still have a 25% probability of being stars)
ukidss = ukidss[(ukidss['mergedClass'] == 0) | (ukidss['mergedClass'] == -1) | (ukidss['mergedClass'] == -2) | (ukidss['mergedClass'] == -3)]
# remove dublicates of sources that are detected in several overlapping fields.
# http://surveys.roe.ac.uk/wsa/www/gloss_p.html#udssource_priorsec
ukidss = ukidss[(ukidss['priOrSec'] == 0) | (ukidss['priOrSec'] == ukidss['framesetID'])]
# There is no second epoch data in L1688
ukidss.remove_columns(['K_2AperMag3', 'K_2AperMag3Err'])
ukidss.rename_column('K_1AperMag3', 'KAperMag3')
ukidss.rename_column('K_1AperMag3Err', 'KAperMag3Err')
# For some reason saturated stars are not flagged in the data we downloaded.
# From comparison with 2MASS we find that sources with either UKIDSS or 2MASS mag < 10 are not to be trusted in UKIDDS.
# In this step I remove all data with mag < 10 from the UKIDSS catalog. Those which are so bright, that they overflow and actually appear fainter are still in the catalog
# and need to be removed in comparison to 2MASS after crossmatching.
for band in 'ZYJHK':
    ind = (ukidss[band+'AperMag3'] < 10.)
    ukidss[band+'AperMag3'][ind] = np.nan
    ukidss[band+'AperMag3Err'][ind] = np.nan
    # also: rename columns to include name of survey
    ukidss.rename_column(band+'AperMag3', 'UKIDSS_'+band+'mag')
    ukidss.rename_column(band+'AperMag3Err', 'e_UKIDSS_'+band+'mag')

ukidss.rename_column('sourceID', 'UKIDSS_sourceID')
ukidss_names = [n for n in ukidss.colnames if 'UKIDSS' in n]

cat.add_catalog_data(ukidss, names = ukidss_names, ra1='ra', dec1='dec', ra2='RA', dec2='Dec')
# now remove those values that are so bright in 2MASS that they should be saturated in UKIDSS
for band in 'JHK':
    ind = cat[band+'mag'] < 10
    cat['UKIDSS_'+band+'mag'][ind] = np.nan
    cat['e_UKIDSS_'+band+'mag'][ind] = np.nan
```

```python
def pick_best_simbad_names(names):
    '''Given a list of strings (SIMBAD identifiers), pick one best and up to three secondary names from SIMBAD for labelling plots etc.'''
    # small comp(lexity) means it's a good name to use
    for i in range(len(names)):
        names[i] = names[i].replace('NAME ', '')
    if len(names) == 0:
        return '',''
    elif len(names) == 1:
        return names[0], ''
    else:
        comp = np.zeros((len(names), 5))
        for i,s in enumerate(names):
            comp[i,0] = len(s)
            # Penalties or rewards if certain well-known numbering schemes appear in any identifier
            comp[i,1] = 5 * ('[' in s) - 5 * ('V*' in s) - 5 * ('ISO-Oph' in s) - 5 * ('2MASS' in s)
        comp = comp.sum(axis=1)
        ind = np.argsort(comp)
        #print list(names)
        return names[ind[0]], ', '.join(names[ind[1:min(4, len(names)-1)]])
```

### Clean the catalog

Some of the very faint sources are artifacts, e.g. PSF spikes that in some frames are found as sources by the source detection algorithm. I inspected all lightcurves of detected variable objects and looked at their cut-out images to identify sources that might be spurious. Then, I asked Robert Gutermuth to generate movies with cut-out images from every single frame where this objects is detected. If turned out that almost all source that look suspicious on one frame do so in each frame. Additionally, I inspected lightcurves with single-point outliers for faint stars. In all cases these turned out to be uncleaned cosmics. Thus, those measurements need to be removed from the data I use here.

The way we do that is that we only keep lightcurves that are matched to a sources in a catalog that has very few artifacts: Gutermuth et al (2009), c2d, or 2MASS.
The UKIDSS catalog contains quite a few artifacts, thus a sources that only has a UKIDSS counterpart will be rejected.

```python
plt.plot(cat['coherence_time_36'], cat['coherence_time_45'],'rs', label = 'Coherence time')
plt.plot(cat['cumsumtime_36'], cat['cumsumtime_45'],'bo', label = 'Cumulative sum')
plt.xlabel('IRAC1')
plt.ylabel('IRAC2')
plt.title('Compare different measures of the ACF')
plt.xlim(0,14)
plt.ylim(0,14)
plt.legend()
```

The plot above compares the measures of the auto-correlation function (ACF) for IRAC1 and IRAC2. If both bands are correlated (some objects might have strong color variations, but that is the exception), then the ACF for both bands should be similar.

The plot compares the coherence time (the width of the first peak in the re-sampled ACF) and the cumulative sum (a measure with a similar idea, but defined for the non-resampled ACF) for IRAC 1 and IRAC 2. For short times (t < 2. days) it's not quite clear that they are related, but for longer times the number obtained from both channels seem to agree to some extend, indicating that the values measured are real and not just noise.

Another way to plot this is to compare the coherence time and the cumulative sum for a single channel and then see if the statistical distributions are similar in both channels.

```python
plt.plot(cat['coherence_time_36'], cat['cumsumtime_36'],'rs', label = 'IRAC 1')
plt.plot(cat['coherence_time_45'], cat['cumsumtime_45'],'bo', label = 'IRAC 2')
plt.xlabel('coherence time')
plt.ylabel('cumsum time')
plt.title('Compare different measures of the ACF')
plt.xlim(0,10)
plt.ylim(0,14)
plt.legend()
```

Ignoring a few outliers with long coherence time and also long cumsum time, the distribution seems anti-correlated as expected. The outlier might be lightcurves with a low signal. We can test that be selecting them and looking at their lightcurves or ACFs visually. Below, I inspect the most extreme examples.

```python
ind = (cat['coherence_time_45'] > 4.) & (cat['cumsumtime_45'] > 4.) & np.isfinite(cat['cumsumtime_45'])
for i in ind.nonzero()[0]:
    t = cat.lclist[i]['t45']
    print '{0} datapoints in lightcurve {1}'.format(len(t[t<55340]), i)
    plt.plot(t, cat.lclist[i]['m45'])
plt.xlim(55290,55340)
ax = plt.gca()
ax.invert_yaxis()
ax.set_xlabel('MJD')
ax.set_ylabel('[4.5]')
```

All those lightcurves are fairly faint and fairly noisy. Also, some of them have long gaps in the data. When building up the ACF these gaps get interpolated linearly, which will cause a long coherence time in the resulting ACF. The cumsumtime uses the original data without any resampling, hence it does not suffer from this problem. It's long because there is no apparent short-term correlation, but also because there are very few data pairs with a small $\delta t$.

So, I conclude that the measures of the non-periodic time scale from the ACF are useful, if either of the time scales is > 2 days (in which case the other one will be short, since they are anti-proportional).

## Introduction

Stars form in dense and cool molecular clouds. When the local density is high enough, the matter can gravitationally collapse and form a young stellar object (YSO). In the early phases the thick envelope dominates the emission from the YSO and hides what is going on within (class~I). Eventually, the envelope flattens out to a circum-stellar disk. This disk still causes an IR excess above the level of a stellar photosphere (class~II or classical T Tauri star - CTTS), which can be used to distinguish those objects from more evolved stars, e.g. using the \emph{Spitzer} InfraRed Array Camera \citep[IRAC,][]{2004ApJS..154...10F} camera. The accretion disk does not reach down to the central star. Instead, the inner edge is truncated by the stellar magnetic field. The inner radius of the optically thick dust in the disk is larger and mostly given by the dust-sublimation temperature. Some of the mass in the circum-stellar disk is accreted onto the central star via magnetically confined accretion funnels, some is blown out by disk winds and some condenses into planets. When the disk is cleared, the IR colors of the YSO match those of main-sequence stars (class~III or weak-lined T Tauri star - WTTS). In addition to the circum-stellar absorption many YSOs are embedded in the molecular cloud, so that even class~III objects can appear reddened. 

TTS were originally identified by their variability \citep{1945ApJ...102..168J} -- long before anybody realized that TTS are indeed pre-main sequence stars. In most cases, particularly in CTTS and WTTS, the dominant time scale in the optical is the stellar rotation period, typically a few days \citep{1983ApJ...267..191R,1986A&A...165..110B,2009ApJ...694L.153N}. YSOs can have cool spots caused by magnetic activity similar to our Sun and also hot spots which mark the impact points of the accretion funnels onto the stellar surface. This impact happens at free-fall velocities up to 500~km~s$^{-1}$, thus the accretion shock heats the accreted mass to X-ray emitting temperatures \citep[e.g. review by][]{2011AN....332..448G}. In the optical, the accretion region appears as emission that often is approximated as a black-body with $10-20\,000$~K (\citealp{calvetgullbring}, but see also \citealp{2012AstL...38..649D,2013AstL...39..389D}). Variability in the mass accretion rate can lead to changes in the hot spot signatures. The dynamical time scale that controls the accretion is the Kepler period of the inner disk where the accretion funnels start and indeed optical variability with amplitudes around 0.1~mag is often observed in CTTS on this time scale \citep{1996MNRAS.282..167S,2008MNRAS.391.1913R}. Another source of variability related to the accretion could be shock oscillations. This has been predicted theoretically \citep[e.g.][]{2008MNRAS.388..357K}, but is not observed so far \citep{2009ApJ...703.1224D,2010A&A...518A..54G}. However, \citet{2011AJ....142..141B} find indications that strong accretion in V1647~Ori could excite radial pulsations of the star itself.

YSOs can also vary on much longer time scales. Variability on the scales of years could be caused by changing circumstellar extinction \citep{2007A&A...461..183G} or by massive accretion events when a significant fraction of the disk mass drains onto the YSO. In this case the accretion luminosity can outshine the YSO by orders of magnitudes and it takes months to years \citep[in the case of EXor outbursts,][]{2012ApJ...749..188L} or even centuries \citep[FUor outbursts,][]{1996ARA&A..34..207H} until the accretion decays back down to the original level. Again, these phenomena are mostly observed in the optical. 

In this article, we present a monitoring campaign of the star forming region L1688 in the mid-IR to compare and contrast the findings of optical and near-IR observations.

The spectral energy distribution (SED) of YSOs in the optical is dominated by the stellar photosphere and the accretion spot. Thus, optical monitoring is very effective to understand the stellar rotation and the accreting spot. The disk radiates mostly at longer wavelengths, which are probed in the observations presented in this article. Depending on the mass of the disk and the size of the inner hole it will start to dominate the SED at $K$ band or in the IRAC bands at $3.6\mu$m and $4.5\mu$m. Simple disk models still treat the disk as a static and axisymmetric structure, but observationally it now seems that the disk is in fact ``a bubbling, boiling, wrinkled, dented, warped mass of gas and dust'' \citep{2013AJ....145...66F}.

This paper is part of the YSOVAR (Young Stellar Object VARiability) project, that has monitored the \object{Orion nebular cluster} and eleven smaller star forming regions with \emph{Spitzer} in the warm mission in $3.6\mu$m and $4.5\mu$m to understand the mid-IR variability of YSOs. First results on the \object{Orion Nebula cluster} (ONC) are published in \citet{2011ApJ...733...50M} and \citet{2012ApJ...753..149M}. More details of the observing strategy and an overview of the data can be found in \citet{luisa} (from now on ``paper~I''). In this article, we present global properties of the mid-IR variability in \object{L1688}.

First, we introduce L1688, the star forming region targeted by our observations (section~\ref{sect:thestarformingregionl1688}).
In section~\ref{sect:observationsdatareductionandauxiliarydata} we introduce the data reduction and discuss source lists and stellar properties obtained from the literature. Section~\ref{sect:classification} classifies all sources according to cluster membership, spectral slope and variability. After that section~\ref{sect:results} discusses the color changes of sources and how these findings depend on the class of the object. Section~\ref{sect:discussion} discusses physical models to explain the observed features in the lightcurves. We end with a summary and some conclusions in section~\ref{sect:summary}.

## The star forming region L1688

Lynds~1688 (L1688) is a sub-cloud of the $\rho$~Oph star forming region, one of the best-studied young clusters in the sky \citep[review by][]{2008hsf2.book..351W}. The distance to the cloud is a matter of active discussion. VLBA measurements and Hipparcos parallaxes point to values around 120~pc \citep{2008ApJ...675L..29L,2008A&A...480..785L}, while \citet{2008AN....329...10M} find 135~pc, again based on Hipparcos data. One possible solution is that the streamers of L1688 are closer to the Sun than the core of the region \citep{2008A&A...480..785L}. In this article, we adopt a distance of 120~pc, but we note that most of our results are independent of the actual distance. The central region of L1688 is very dense and deeply embedded ($A_V=50-100$~mag, see figure~\ref{fig:map}), thus no complete survey of all cluster members exists, but an extinction limited, spectroscopic survey \citep{2011AJ....142..140E} finds an average age of 3.1~Myr for a 6.8~pc$^2$ region centered on L1688 and no significant deviation from the interstellar mass function. Earlier studies \citep{1995ApJ...450..233G,1999ApJ...525..440L, 2002A&A...393..597N} concentrated on the deeper embedded core and found a much younger age of 0.3~Myr. Some, but not all of this difference is due to the specific reddening laws or pre-main sequence evolutionary tracks used in these studies \citep[see discussion in][]{2011AJ....142..140E}.

\citet{2012PASP..124.1137F} present spectroscopic monitoring for five YSOs in L1688 near-simultaneous to the \emph{Spitzer} observations discussed here. They do not see any correlation between the hydrogen emission lines that are accretion indicators and the features in the IR lightcurves of their targets indicating that the relatively modest variability in the observation period is not caused by changes in the accretion rate. Notable objects with well-sampled NIR lightcurves are \object{WL 4} \citep{2008ApJ...684L..37P} and \object{YLW 16A} \citep{2013A&A...554A.110P} which show eclipses with periods of 131 and 93~days, respectively. These sources can be interpreted as multiple systems, where one or more components are eclipsed by a warped circumstellar disk. Thus far, the most comprehensive study of IR variability of YSOs in L1688 is \citet{2013arXiv1309.5300P}, who made use of a 2MASS \citep{2006AJ....131.1163S} calibration field that overlaps L1688 so each source has up to 1584 datapoints in $J$, $H$, and $K_s$ spanning 2.5~years. They found 79\% of the known YSOs to be variable. 32 sources are periodic (including cool starspots, hot accretion spots and 6 systems with eclipses), 31 sources show a long-term trend and 40 source vary aperiodically on shorter time scales. The new data presented in this article complements the \citet{2013arXiv1309.5300P} study with observations at longer wavelength.

Three fields in L1688 were observed with \emph{Spitzer} in four observing periods from 2010-04-12 to 2010-05-16, 2010-09-22 to 2010-10-27, 2011-04-20 to 2011-05-23 and 2011-10-01 to 2011-11-06. These periods are consecutive visibility periods dictated by the \emph{Spitzer} orbit \citep{2004ApJS..154....1W}. In the first observing period the sampling is much denser. About 70 epochs have been observed on most sources with irregular time intervals to aid period detection and avoid aliasing with the observation frequency in contrast to ground-based data that is frequently effected by an alias due to the day-night cycle on the earth. In the first window, we used a repeating pattern every 3.5 days for easy 1-week scheduling.  Within the 3.5 day window the time-step increased by two hours starting with 2 hours.  In the last three visible windows, the time steps were linearly increasing of 1, 2, 3 days etc.\ to again sample multiple variability frequencies equally. Less than ten datapoints per source were taken in each of the later three observation periods. In total there are 108 epochs of observation with a total exposure time of 30.7 hours.  Figure~\ref{fig:cadence} shows the distribution of the epochs. They can be found under Program Identification number (PID) 61024 in the Spitzer Heritage Archive. Each epoch consists of six exposures in IRAC mapping mode using the high-dynamic range (HDR) data acquisition, which obtains an 0.4~s and an 10.4~s exposure for each pointing. 

The three fields chosen were observed with the IRAC1 and IRAC2 channels (effective wavelengths 3.6$\mu$m and 4.5$\mu$m). Both channels operate simultaneously, but their fields-of-view are non-overlapping. Thus, each target field is observed in two consecutive pointings, one for each channel. In each observation of one of the target fields we automatically obtained data from a neighboring field observed by the channel that is offset from the target field. In the April to May monitoring blocks sources south of the main fields only have IRAC1 data, while those to the north only have IRAC2 data. In the September to November monitoring blocks the situation is reversed.  Not all sources in the central fields have usable data in both bands, because they might be too bright or too faint in one channel. Additional sources with two band coverage are found where the northern side field of one target field overlaps with the southern side field of another target field.

A very detailed account of the data processing and the source extraction is given in paper~I. Here we summarize the main data reduction and processing steps. Basic calibrated data (BCD) are obtained from the \emph{Spitzer} archive. Further data reduction is performed with the IDL package cluster grinder \citep{2009ApJS..184...18G}, that treats each BCD image for bright source artifacts. Aperture photometry is run on individual BCDs with an aperture radius of $2\farcs4$. To increase the signal-to noise ratio and to reject cosmic rays, the photometry from all BCDs in one epoch is combined. The reported value is the average brightness of all BCDs within that epoch that contain the source in question, after rejecting outliers. The photometric uncertainties obtained from the aperture photometry are, particularly for faint sources, only lower limits to the total uncertainty, since distributed nebulosity often found in star forming regions can contribute to the noise. To improve these estimates paper~I introduces an error floor value that is added in quadrature to the uncertainties of individual photometric points. The value of the error floor is 0.01~mag for IRAC1 and 0.007~mag for IRAC2.

We cross-match sources from individual observations with a matching radius of $1\arcsec$ with each other and with the 2MASS catalog, which is used as a coordinate reference. All photometric measurements performed in the context of the YSOVAR project are collected in a central database, which we intend to deliver to the IRSA for general distribution. 
Data for this article was retrieved from the YSOVAR database on 31-Oct-2013 and further processed using custom routines in python. 
We retain all sources that have at least five datapoints in either IRAC1 or IRAC2. Figure \ref{fig:map} shows the distribution of the sources with lightcurves in our input catalog overlayed on a larger reverse-color IRAC1 map observed during the cold mission.

#### Instrumental effects in IRAC lightcurves

```python
def characterize_seasonal_jump(t, m, merror):
    '''Characterize the seasonal jump (an instrumental effect)

    This is +1, is the mean in season 1 and 3 is higher than the mean in season 2 and 4,
    -1 if it is vice-versa and 0 if the mean values for each season do not correlate with the 
    time of year the obervations were taken.
    ``np.nan`` is returned for lightcurves that do not have data in all seasons.
    '''
    seasons = np.array([[55295,55338], [55458,55500], [55668, 55710], [55830, 55890]])
    mean = np.zeros(4)
    n = np.zeros(4)
    for i in range(4):
        ind = (t > seasons[i,0]) & (t < seasons[i,1])
        mean[i] = np.mean(m[ind])
    ind = np.argsort(mean)
    if (set(ind[0:2]) == set([0,2])) & (np.min(mean) > 0):   # the mean > 0 makes sure that data exisits in all seasons.
        return 1
    elif (set(ind[0:2]) == set([1,3])) & (np.min(mean) > 0): 
        return -1
    elif (np.min(mean) > 0):
        return 0
    else:
        return np.nan
    
YSOVAR.registry.register(characterize_seasonal_jump, error = True, time=True, force=True, name='seasonaljump')
```

Despite the careful data reduction described above, some residual artifacts remain in the \emph{Spitzer} lightcurves. In this section, we search for artifacts that are related to the position on the detector. Compared with other clusters in the YSOVAR project L1688 is particularly suited to discover those since the fields observed in L1688 have almost no rotation within one season so that each source falls onto the same region on the chip. The spacecraft orientation flips between seasons, so that most instrumental artifacts produce lightcurves that have one baseline level in season one and three and another in season two and four. Figure~\ref{fig:instrumentallcs} shows examples of lightcurves that might fall into this category.
\begin{figure}
\plotone{instrumentallcs}
\caption{Lightcurves with apparent jumps between seasons. The examples shown here are all $4.5\mu$m lightcurves, but the same can be seen in the $3.6\mu$m channel for
some sources. 
The mean brightness of these objects is (from top to bottom) 8.2, 14.2, 14.6 and 14.8~mag.
\label{fig:instrumentallcs}}
\end{figure}

Sufficiently strong intrinsic variability can mask this effect and for any individual source this instrumental effect cannot be distinguished from intrinsic variability with a period of one year. Thus, we use a statistical approach to quantify the number of lightcurves that suffer from this problem. For each channel we have about 200 lightcurves with datapoints in all four observing seasons (the total number of lightcurves is larger, but only for sources in the three target fields we have data for all four seasons). If the mean magnitude in a season depends on the detector position of the source, then this source will on average be brighter in season 1 and 3 and fainter in in season 2 and 4 (or vice-versa). We calculate the mean magnitude in each season and test weather the two brightest mean magnitudes belong to season [1,3] or [2,4] or any other combination ([1,2], [1,4], [2,3], [3,4]). In this section we will use the following short notation: When we calculate the mean magnitudes in each season for a sample of $a+b+c$ sources, then $a$:$b$:$c$ means that $a$ sources have the brightest mean magnitudes in season [1,3], $b$ sources in season [2,4] and $c$ sources have their two brightest mean magnitudes in any other combination of seasons ([1,2], [1,4], [2,3], [3,4]). If the difference in the mean magnitude between seasons is unrelated to the position on the detector, we expect the ratio $\frac{1}{6}:\frac{1}{6}:\frac{2}{3}$. In contrast, the observed lightcurves have the ratio 47:66:75 for IRAC1 and 52:59:95 for IRAC2. This is incompatible with the expected multinomial distribution (the probability to observe a distribution at least as far from the expected $\frac{1}{6}:\frac{1}{6}:\frac{2}{3}$ by chance is $<10^{-8}$). Comparing the expected and the actual distributions, we see that in each band 50 lightcurves, a quarter of the sample, are missing from the last bin. This implies that about a quarter of all lightcurves suffers from the artifacts described above. However, both IRAC channels use independent detectors and thus the chance that both channels are effected in the same source is only $0.25^2=6$\%, the chance the both channels are effected in the same way (bright-faint, vs. faint-bright) is half of that (3\%).

In section~\ref{sect:variability} we determine the variable members using a Stetson and a $\chi^2$ test as defined in paper~I. The limits in those tests are designed to be conservative and indeed we find a distribution of 5:7:24 for IRAC1 and 5:5:27 for IRAC 2, when we restrict the sample to those lightcurves that will be classified as variable below. Both ratios are fully compatible with the expected multinomial distribution $\frac{1}{6}:\frac{1}{6}:\frac{2}{3}$. This shows that the limits we apply are conservative enough that the sample of variable stars only contains sources where the variability is dominated by the intrinsic source variability.

We visually inspected every frame for a sample of sources with lightcurves similar to those in figure~\ref{fig:instrumentallcs} to identify the cause for the artifacts. We found that several different effects can cause the apparent steps in the lightcurve. Some sources are close to a detector edge, so that the background is not well-determined, some are in the wings of the point-spread function (PSF) of a bright neighbor and some show residuals from hot pixels. There are other sources which appear perfectly normal when looking at the frames. Most of these effects would only affect one of the spacecraft orientations; even the PSF wings change with the orientation since the PSF is not perfectly circular. Figure~\ref{fig:instrumentallcs} shows sources with a magnitude between 8 and 15 in $4.5\mu$. The relative difference in magnitude is smaller for brighter sources, but since the photometric errors also are smaller these instrumental effects can still be significant. 

In summary, about a quarter of all lightcurves is effected by detector position dependent artifacts. In extreme cases, these reach 0.5~mag for 14~mag source and up to 0.05~mag in bright sources around 8~mag. However, we show statistically, that the definition of variability we use is so conservative, that this effect does not contribute to our sample of variable sources/ 

Furthermore, we visually checked all frames for lightcurves that are classified as variable in section~\ref{sect:variability} and removed datapoints visibly effected by instrumental artifacts (cosmic rays, read-out streaks for bright neighbors).

Just over 10\% of the 884 sources are classified by \citet{2009ApJS..184...18G}. Of those sources, 16 are class~I sources, 36 are class~II sources and 59 are class~III candidates, i.e.\ disk-less sources, which either lost their disks only recently or they are main-sequence stars. The \citet{2009ApJS..184...18G} classification scheme is conservative, it represents a set of sources that can be dereddened and classified to a high degree of confidence. However, for many sources we lack the required spectral coverage. Thus, paper~I defines a simpler approach, which uses the observed colors only. In this scheme, we fit the spectral slope $\alpha=\frac{d \log \lambda F_{\lambda}}{d \log \lambda}$, where $\lambda$ is the wavelength and $F_{\lambda}$ the flux per unit wavelength interval at that wavelength. We make use of all measured fluxes (no upper limits) in the 2-24$\mu$m range, which corresponds to the range from the $K$ filter, which we take from 2MASS and/or UKIDSS, to the MIPS 24$\mu$m channel. We use the fluxes for all \emph{Spitzer} bands from \citet{2009ApJS..184...18G} or c2dr and summarize the new lightcurves by calculating the mean and standard deviation for the new IRAC1 and IRAC2 data. Thus, sources detected both in the cold mission and in the new dataset, will have two datapoints for IRAC1 and IRAC2. We perform a least-squares fit and call sources with $0.3 < \alpha$ class~I, $-0.3 < \alpha <0.3$ flat-spectrum (F), $-1.6<\alpha < -0.3$ class~II sources and $\alpha < -1.6$ class~III candidates. In this scheme a significantly reddened main-sequence star, which has an intrinsic slope $\alpha<-1.6$, may appear as class~II, flat-spectrum or even class~I object, so this observational classification cannot be translated directly into the evolutionary stage of an object. We find 110 class~I, 78 flat-spectrum and 455 class~II sources and 221 class~III candidates. 20 sources cannot be classified, because they are seen in one band only. Comparing with \citet{2009ApJS..184...18G}, the resulting classification is very similar, particularly for the class~I and flat-spectrum sources. Most sources with a spectral slope $<0.3$ according to our slope-fitting are class~I sources in both classification schemes. With two exceptions our flat-spectrum sources are either class~I or II in \citet{2009ApJS..184...18G}. Of the 52 sources with a spectral slope between -1.6 and -0.3, 21 are also called class~II by \citet{2009ApJS..184...18G}, but 30 are class~III in that paper, indicating that these may be reddened background stars. The general agreement between the more complex classification scheme and the observed spectral slope is also found for other star forming regions (paper~I).

We use three different methods to detect variability in all lightcurves independent of their membership status or SED slope.

#### Stetson index

 For details on those methods see paper~I. Sources in the target fields are observed in IRAC1 and IRAC2 within a few minutes and we calculate the Stetson index $s$ with points weighted evenly \citep{1996PASP..108..851S}.
\begin{equation}
s = \frac{1}{\sqrt{N(N-1)}}\sum_{k=1}^{N}sig(P_k)\sqrt{|P_k|}
\label{eqn:stetson}
\end{equation}
where the sum is taken over all $N$ pairs of observations in IRAC1 and IRAC2 with observed magnitudes $a_k$ and $b_k$ and uncertainties $\sigma_{a_k}$ and $\sigma_{b_k}$. $sig$ denotes the sign of $P_k$. $P_k$ is the product of the normalized residuals in both bands:
\begin{equation}
P_k  = (\frac{a_k-\bar{a}}{\sigma_{a_k}}) \times (\frac{b_k-\bar{b}}{\sigma_{b_k}})
\end{equation}
Here $\bar{a}$ denotes the error weighted mean of all IRAC1 magnitudes and $\bar{b}$ the error-weighted mean of all IRAC2 magnitudes. The Stetson index is very robust to observational errors since those are unlikely to effect both bands in the same way. Following paper~I, we define a source as variable if $s>0.9$. Figure \ref{fig:stetsonlcs} shows examples of lightcurves that are classified as variable according to the Stetson index. In the first season we find 34 sources to be variable, that is 18\% of all lightcurves where we can calculate the Stetson index. The number is similar (39 sources, 18\%) if we consider all observing seasons. This is not surprising, given that the Stetson index is designed to be a ``robust'' statistic that is little influenced by a small number of datapoints and compared to the first season, the other seasons only contribute a small number of datapoints. We caution that equation~\ref{eqn:stetson} will introduce a systematic bias in such a way that a source that is variable in only one band will not be marked as variable.

For sources outside the primary target fields or for sources which are too bright or too faint in one IRAC channel the Stetson index cannot be calculated. For those sources we calculate the reduced $\chi^2_{red}$ to the mean of the lightcurve:
\begin{equation}
\chi^2_{red} = \frac{1}{N-1}\sum_{k=1}^{N}\frac{(a_k-\bar{a})}{\sigma_k^2}
\end{equation}
Instrumental uncertainties lead to a non-Gaussian error distribution and we use a conservative cut-off and mark sources as variable only if $\chi^2_{red} > 5$ (paper~I). Using this metric, we find 22 sources that exhibit variability in IRAC1 and 18 that exhibit variability in IRAC2. Examples are shown in figure~\ref{fig:chi2lc}. If we fit a linear slope instead of comparing to the mean (i.e.\ a constant), no source is classified differently. 
%When fitting higher order polynomials, the number of sources with $\chi^2_{red} > 5$ goes down, e.g. for a quadratic polynomial only 29 and %18 sources have a $\chi^2_{red}> 5 $ in IRAC1 and IRAC2, respectively. 
Many sources have significantly larger values of $\chi^2_{red}$ for the whole observation series than within one season indicating a constant luminosity over 40~days, but change it over time scales between six months and two years.

Last, we search for periodicity in the fast-cadence (season 1) lightcurves using a Lomb-Scargle periodogram \citep{1976Ap&SS..39..447L,1982ApJ...263..835S}. We require a false alarm probability $<0.03$ and to further reduce the number of false positive, we additionally require that at least one of three supplementary algorithms that we run points to a periodicity with the similar timescale (see paper~I for details). We only search for periods between 0.1 and 14.5 days, so that about three periods fit in each observing season. Finally, all algorithmically detected periodicity is vetted by eye. If periods are found in multiple bands, we report the period in IRAC1, which is generally the most reliable, if IRAC1 did not turn up a periodicity we report the value for the IRAC2 lightcurve and as a last resort the period found in the IRAC1-IRAC2 color. The final list of adopted periods is given in table~\ref{tab:period}. It contains sources from all SED classes.

Allowing longer periods leads to the detection of many long-term trends, where the data does not show that these trends are actually periodic. We find 14 sources that show a periodic behavior in the first observing season, but in all cases this periodicity is not stable over more than one season. Ten of those sources have information in both IRAC bands and eight are already classified as variable by the Stetson index, and three out of four sources with data in one band only are variable according to the $\chi^2_{red}$ test. The number of datapoints in seasons 2, 3, and 4 is too low to search for periodicity within an particular season. Figure~\ref{fig:periodlc} shows the phase-folded lightcurve in one band for six periodic sources where the periodicity is significant in a single band. For one source, the period is seen only in the color term [3.6]-[4.5]. The folded single band and color lightcurves for this source are shown in figure~\ref{fig:periodcolorlc}. Most of the periods are in the range 3-7 days, only two sources have longer periods of 11 and 14 days. The largest amplitude is around 0.3~mag, the smallest around 0.05~mag. The datapoints taken in the first season follow the folded period much better than the remaining data, indicating that the period is not stable for longer than a few months. In ROXN~41 the datapoints of the later seasons, folded with the same period, seem to follow a different, yet clearly defined lightcurve with a similar period. However, due to the low number of points, this is statistically less significant.

#### Summary of variability detection


The largest group of lightcurves in the sample of L1688 does not have any significant variability. The lower limit where variability is detected depends mostly on the brightness of the object - fainter sources need a stronger relative variability due to the larger measurement uncertainties. Paper~I presents detailed Monte-Carlo simulations to show that the Stetson test finds variability, when the amplitude is a few times larger than the average uncertainty. The exact number depends on the signal shape. For example, the variability in a source that switches between two discrete levels will be detected, if the amplitude is at least twice the uncertainty (0.03~mag for a star with magnitude 14) if the source is in each state about half the time. If data from only one band is available, the step size must be more than four times larger than the uncertainty to be found by the $\chi^2$-test (0.06~mag for a star with magnitude 14.). Given the observing cadence the Monte-Carlo simulations show that we are sensitive to periods between 0.1 and 14.5 days (paper~I).

Figure~\ref{fig:probvarSED} shows the fraction of variable stars for cluster members and non-members using the YSOVAR cluster membership criterion from section~\ref{sect:membership}. Section~\ref{sect:variability} describes the criteria for variability in detail. Sources are considered variable, if they have a low false alarm probability for periodicity or a Stetson index $>0.9$ (if simultaneous data in IRAC1 and IRAC2 exists) or $\chi^2_{red}>5$ (if no simultaneous data exists). Note that the classes shown here are based on the observed SED and thus background non-member stars seen through the cloud might appear red (class II) even if their intrinsic SED slope is compatible with SED class~III. No star classified as a member is fainter than 14~mag. 

The observed fraction of variable stars is the best-guess for the probability to find variability for a star of this SED class. We calculate the uncertainty for the probability based on the observed number of variable stars $v$ in a bin with $N$ total sources as follows: If the true probability for a source in the bin to be variable is $p$, then the probability to observe $v$ variables in $N$ stars is given by $p_{obs}$, which follows a binomial distribution:
\begin{equation}
p_{obs}(v,N,p) = {N \choose v} p^v (1-p)^{N-v}
\end{equation}
The peak of this distribution is $p_{obs}(p) = \frac{v}{N}$. We then calculate the boundaries of the confidence interval $p_1 .. p_2$ for $p$ such that $p_{obs}(p_1) = p_{obs}(p_2)$ and the area under the curve includes 68\% of the probability. Those uncertainty ranges are shown in figure~\ref{fig:probvarSED}. 

The sample of faint stars (green bars) has almost the same fraction of variable sources in every SED class. The low fraction can be explained by two effects. First, variability of faint sources cannot be reliably detected unless the amplitude is exceptionally large. Second, as most cluster members are brighter than 14~mag, the faint sample is less contaminated by unrecognized variable cluster members than the sample of brighter non-member sources.
The observed probability for all classes of bright non-members is compatible with a variability fraction just under 10\%. The fact that this barely changes with the SED class, much unlike the distribution for the cluster members, indicates that the contamination with unidentified cluster members is low since they would bias the observed variability fraction to higher values for the younger classes. The vastly different number of sources in the different bins make it highly unlikely, that each bin is contaminated by the same fraction of class~I-II cluster members.

All known cluster members in class~I are variable. This fraction decreases to about 80\% in class~II sources and then drops rapidly. In class~III sources it is already compatible with the fraction observed for non-members. This distribution is a powerful indication that the main cause for the variability must be found in the inner disk regions. 
In principle the SED classes of YSOs (class~I, flat-spectrum, II and III) represent an evolutionary sequence. Initially, the entire emission from the YSO is dominated by the dusty envelope (class~I and flat-spectrum). As the system evolves the dust settles in a disk (class~II). Later the disk disperses entirely due to planet formation, accretion and evaporation, so that all IR emission is photospheric. 

In class~I and flat-spectrum sources a significant fraction of the observed IR radiation comes from the envelope, but the variability is still present when the envelope disperses and the mass concentrated in the disk (class~II). However, when the inner disk vanishes, the variability suddenly becomes undetectable. One complication is that intermediate dust, that is not associated with the specific YSO, also reddens the spectrum and makes the star appear younger in an SED evolutionary sense than it actually is, e.g.\ a background star that happens to be found behind a thin layer of the cloud might appear as class~II or even flat-spectrum source based on infrared SED analysis alone. Also, the orientation of the specific system plays in important role. Already at very early stages (class~I) YSOs can develop powerful outflows, which pierce their envelope and allow us to view the star through the outflow cavity, so that the photospheric contribution becomes important in an early stage. On the other hand, if we see the accretion disk edge-on it will cause a high degree of reddening even for late evolutionary stages, when the envelope is already dispersed. Simulations show that stars of very different ages can thus exhibit a similar SED \citep{2006ApJS..167..256R}. These effects tend to smooth out age effects when comparing different evolutionary classes, as each bin will contain a few sources of neighboring evolutionary stages.
Similarly, the membership lists for L1688 might not be complete, so that some sources that we treat as non-members here are actually YSOs.

Figure~\ref{fig:meanampSED} shows the median amplitude of the observed variability for all variable sources $<14$~mag. The errorbars in this figure represent the standard deviation of the mean amplitudes within one class. 
For members the variability is stronger in class~I and flat-spectrum sources than in class~II sources, where the amplitude is still larger than in class~III sources. Also, the spread of the amplitudes in the earlier classes is much larger than for the later classes. For non-members the observed amplitudes seem to follow a similar, but less pronounced trend. This suggests that this category includes not only reddened background objects with an intrinsic class~III SED, but also some unidentified cluster members.

In summary, we can say that the probability for a source to be observed as a mid-IR variable decreases only a little between class~I and II sources, but the mean amplitude as well as the differences within a class are much larger for younger sources which have more circumstellar material.

As the last aspect in this section we will look at the timing behavior of the lightcurves. The list of periodic sources contains objects of all evolutionary stages, but is too small to recognize any trends in the period with evolutionary stage. To asses changes in the time-scale in all sources, not just the periodic ones, we calculate the discrete auto-correlation function $ACF(\tau)$ for each fast-cadence (season 1) lightcurve. One complication here is that our data is unevenly sampled. In order to calculate the $ACF$ we linearly interpolate the lightcurve on a grid with time steps of 0.1~days. This process can change the properties of the lightcurves on short time scales. However, for most lightcurves the relevant time scales are longer than the distance between two observations (0.1 to 0.8 days). We use the following definition of the $ACF$
\begin{equation}
ACF(\tau) = \frac{1}{N-\tau}\sum_{k=0}^{N-\tau}\frac{(a_k-\bar{a})(a_{k+\tau}-\bar{a})}{\sigma_a^2}
\end{equation}
where $N$ is the total number of points in the lightcurve $a_1, a_2, ... , a_N$ with a mean of $\bar{a}$ and a standard deviation of $\sigma_a$. In the discrete $ACF$ $\tau$ is the number of timesteps.
\begin{figure}
\plotone{ACF}
\caption{Histrogram of the time scale $\tau$ in the auto-correlation function (ACF) for variable sources in different SED classes, such that $\tau$ is the smallest tine scale with $ACF(\tau)<0.5$.
\label{fig:ACF}}
\end{figure}
By definition $ACF(\tau=0)=1$ and the $ACF$ decays for longer time-lags $\tau$. We take the first value of $\tau$ with $ACF(\tau)<0.5$ as the characteristic time-scale for a lightcurve. A more common definition would be to use the position of the first local maximum in the $ACF$, but due to the low number of datapoints in our lightcurves, the noise in the auto-correlation function is large and this value is often not well defined. While our definition might not give us the time scale of the physical processes at work, they are still valid for relative comparisons. We find that the average $ACF$ time scale decreases from 4~days for class I sources to 2.5 days for class~III sources. A more detailed analysis of stochastic and quasi-periodic properties of the lightcurves is beyond the scope of the current paper \citep[see][for a discussion of these properties in $JHK$ lightcurves]{2013arXiv1309.5300P}.

We can recognize different morphological types of lightcurves. The largest group are lightcurves with no significant variability. Those lightcurves are not analyzed any further in this work. The remaining lightcurves can be broadly grouped into four categories: Periodic, quasi-periodic, aperiodic and long-term trend lightcurves. Periodic lightcurves are described in section~\ref{sect:variability} and examples can be seen in figure~\ref{fig:periodlc}. The other types are described qualitatively here; their classification will be quantified in forthcoming publications of the YSOVAR project.

We call a lightcurve quasi-periodic when no significant period is found, but by eye we can see that the up and down has a preferred time scale. Examples are shown in figure~\ref{fig:morelcs} (left column). \object{WL 3} shows variability with a time scale $\sim3$~days in the first half of the first observing season, but around MJD 55316 the flux increases significantly and this rise dominates over any periodicity. The dip around MJD~55325 again has a similar duration as those observed in the beginning of the season. The other two sources shown in the left column of the figure might be similar, but the signal is not as good.

The morphology of the lightcurves noted here in the IRAC1 and IRAC2 bands echo the results discussed for morphology observed in $JHK$ by \citet{2013ApJ...773..145W}.  They noted that lightcurves of their $\sim$~50 stars could be divided into four roughly equal size groups, which they called periodic, quasi-periodic, long term and stochastic. Here we find a similar division, 10 of 52 members are strongly periodic. Like \citet{2013ApJ...773..145W} we find that even strongly periodic sources are not perfect clocks in the mid-IR in that the amplitude varies from cycle to cycle.  It is also clear that the phase shifts from epoch to epoch. In a very similar study to the one we present here, \citet{2013AJ....145...66F} observed IC 348 over a 40 day window with IRAC1 and 2 and find 25\% of the variable stars are likely periodic.  \citet{2013AJ....145...66F} go on to argue that the remainder are aperiodic (i.e.\ have no intrinsic timescale). Lightcurves similar to what we call quasi-periodic, that is with an identifiable timescale of a few days but no strict periodicity were also seen by \citet{2013ApJ...773..145W}.

For sources in the primary target fields the observations in $3.6\mu$m and $4.5\mu$m are separated by a few minutes only. For those sources we compare the color and the magnitude in the $3.6\mu$m band in a color-magnitude diagram (CMD), see figures~\ref{fig:morelcs} and \ref{fig:cmd}. In some cases, the color is fairly stable within one season, but changes between seasons; in others the time scale of variability is much shorter and color changes are seen within one season as well. For non-variable sources the CMD forms a point cloud with the size set by the photometric uncertainties, but for variable sources the shape of the CMD can reveal the physical cause of the variability, e.g.\ if a disk warp or accretion funnel passes in front of a YSO we expect its color to become redder as it becomes fainter. If the absorber has the same gas and dust properties as ISM, then the datapoints in the CMD follow a line with the slope of the interstellar reddening law. We fit a straight line to all points in the CMD using an orthogonal distance regression method that takes the errors in both the x and y directions into account. From a total of 42 variable sources with CMDs with at least ten datapoints, 37 can be fitted with a formal uncertainty on the best fit slope below $6^{\circ}$. This cut will remove sources without a well-defined slope from the sample (both noise-dominated sources and those where different physical processes contribute to the observed slope). The CMD slopes of sources with a well-defined slope are shown in Figure~\ref{fig:cmdslope}. Sources with uncorrelated noise in both bands would show a reddening slope of $45^{\circ}$, since the x and the y axis in the CMD are correlated, because they both depend on the $3.6\mu$m magnitude. No source is seen in this region.

As can be seen in figure~\ref{fig:cmd} even in those sources with a well defined slope in the CMD the scatter around the best linear fit is not compatible with just the measurement uncertainty: Individual seasons are systematically above or below the line. Note that the seasonal variation in the upper left panel in figure~\ref{fig:cmd} cannot be an instrumental artifact as discussed in section~\ref{sect:instrumentaleffectsiniraclightcurves} because this would cause season two (cyan) and four (red) to be both brighter or both fainter in contrast to the data.

The slopes in figure~\ref{fig:cmdslope} fall in two groups. One group becomes redder when the sources are fainter, because it has slopes comparable to the ISM reddening. This group contains almost all variables that are not classified as cluster members according to the YSOVAR standard membership criterion (section~\ref{sect:membership}). There is a range of evolutionary stages in this group, but most sources have class~I or flat-spectrum SEDs. There is some scatter around the ISM reddening slope indicating that either the reddening law differs between different YSOs or we observe a combination of different physical processes here. The first possibility seems unlikely because in general the reddening in the mid-IR differs only little between different sightlines \citep{2005ApJ...619..931I}. Sources in the other group become bluer when they are fainter. This group consists mostly of class~II and some-flat spectrum sources. Figure~\ref{fig:morelcs} shows the lightcurves and CMDs for two of the bluening sources, \object{WL~3} and \object{[GY92] 264}.

The bottom panel in figure~\ref{fig:cmdslope} quantifies the magnitude of the reddening or bluening observed in each source. In each CMD we project all datapoints onto the best-fit line and then calculate the distance in magnitudes between the 10\% quantile and the 90\% quantile. For a source where the slope in the CMD is compatible with ISM reddening, this can be interpreted as the $A_{3.6}$ of the intervening material. A range of values is observed, but in most sources it is $<0.3$~mag. No significant difference is seen between sources that become bluer or redder when they dim.
We discuss the physical mechanisms that could cause these slopes in section~\ref{sect:colorchanges}.

It has a class~II SED slope and is a YSO of low mass \citep[spectral type M1;][]{1999ApJ...525..440L}. There is no single slope in the CMD, instead the source seems to switch between different modes. Initially, the luminosity drops sharply, while the color is almost unchanged or becomes even slightly bluer. The behavior changes around MJD 55305. For the next 15 days, the source reddens as it dimes, but the the slope is less steep than the ISM reddening law. The evolution is not monotonous, but a brightening (e.g.\ around MJD 55313) corresponds to a bluer color and the source reddens as it dims. At the end of this period, the brightness increases again sharply and the slope in the CMD is comparable to the first period: The source becomes noticeably redder with higher luminosity. In the last 10 days of the monitoring the source again has a slope close to the ISM reddening law. Its luminosity continues to increase and the color becomes bluer until it has a similar color and luminosity as it did at the beginning of the monitoring. This closes the circle in the CMD. This is the only source in our sample where we observe multiple switches of the reddening slope over time. In this case variability similar to ISM reddening corresponds to slow changes in luminosity, while the other direction of the slope in the CMD is associated with faster luminosity changes.

One could thus suspect that sources that follow the ISM slope for a given time have a lower amplitude of variability than sources with a slope that is perpendicular in the CMD. However, comparing the amplitudes of all sources in figure~\ref{fig:cmdslope} that either redden or bluen over the full time span of the observations we do not find significant differences in value or distribution of the variability amplitudes.

In section~\ref{sect:variability} we show that most YSOs (class~I to II) in the L1688 star forming region are detected to be variable. The variability fraction drops from 100\% of all class~I sources with more than five usable datapoints to about 80\% for class~II sources. The expected variability in objects not associated with the star forming region is only a few percent as seen in non-member class~III sources. In the optical and near-IR the fraction of variable field stars is even lower \citep{2009A&A...503..651P,2012A&A...537A.116P,2013AJ....145..113W}. Sources with a class~III SED are not contaminated by disk-bearing stars, since reddening by the cloud can only increase the spectral slope $\alpha$ but never hide an existing IR excess. The larger fraction of non-member variable objects in the other classes in figure~\ref{fig:probvarSED}, thus shows that the membership lists are incomplete. Indeed, we propose that all variable class~I, F and II sources are members and table~\ref{tab:newmembers} lists those new members that are neither contained in our YSOVAR standard membership set (see table~\ref{tab:tab2}) nor in \citet{2008hsf2.book..351W}.

By definition, sources with a class~III SED do not have an IR excess over stellar photospheric colors, thus they are not expected to harbor a disk and cannot show any disk related variability. However, even field stars can show significant variability, at least in the near-IR. \citet{2013AJ....145..113W} find 1.6\% of all field stars to be variable in $JHK$. This includes eclipsing systems, stars with unusually large photospheric spots and other unidentified variables. All effects that influence the $JHK$ lightcurve are likely also visible in the mid-IR. Our fraction of variable objects with a class~III SED in our dataset is higher, but still compatible with the 1.6\% found by \citet{2013AJ....145..113W}. Additionally, inhomogeneities in the intervening cloud L1688 could cause a stochastic, time-variable extinction of background objects. It is consistent with this hypotheses that most variable sources with a class~III SED show only a small scale reddening effect (figure~\ref{fig:cmdslope}). However, only a comparison with mid-IR monitoring of field stars with no cloud in the line-of-sight can show if this effect contributes to the observed variability.

\emph{Spitzer} observations of other young clusters also find that younger stars are more variable, but with lower variability fraction. \citet{2011ApJ...733...50M} find only 70\% of all stars with disks in their ONC sample to have detectable variability; \citet{2013AJ....145...66F} find 60\% in IC~348. Both clusters are located at larger distances and thus observations are less sensitive to small changes. In L1688 a K5 star with little extinction will have a magnitude of 12.2 in $3.6\mu$m using the evolutionary tracks of \citet{2000A&A...358..593S} and the YSO colors from \citet{2013arXiv1307.2657P}. In this magnitude range variability down to 0.05~mag can be detected (paper~I). In contrast, for the same star in the ONC the variability has to be three times larger to be detected. Figure~\ref{fig:meanampSED} shows that there is a wide distribution of amplitudes in YSOs of early classes. Of course, even our observations miss the faint end of the distribution. Comparing our results with the previous studies it seems likely that essentially all YSOs show substantial mid-IR variability on time scales of days to weeks.

Figure~\ref{fig:cmdslope} shows that the observed slopes in the CMD fall into two categories: Sources either become redder as they get fainter or bluer. The reddening group has slopes that are roughly compatible with the slope of an interstellar reddening law. Small deviations can be explained by modifications of the dust-to-gas ratio or the grain size and composition. This is commonly observed in individual young stars \citep{2008A&A...481..735G} and star forming regions as a whole \citep{2010AJ....140..266W,2012AJ....144..101G}. The most prominent example of a class~II source where the extinction changes on time scales of days is \object{AA Tau}. In this case, the absorption dip is periodic and caused by an inner disk warp partially occulting the disk of the star \citep{1999A&A...349..619B,2007A&A...462L..41S}. In the last two years the behavior of AA~Tau changed. It is still variable, but less regular and it appears much fainter. \citet{2013arXiv1304.1487B} explain this with an optical reddening that increased by four magnitudes, most likely due to an asymmetry in the disk that rotated into the line-of-sight. We also observe additional reddening on time scales of months to years in about half of all sources with a well-defined slope in the CMD. An example of that is shown in the top right panel of figure~\ref{fig:cmd}. We do not know the inclination of the objects in our sample, but if it is close to edge-on, this might be caused by an asymmetric disk similar to AA~Tau. The required warp or local change in scale height, that lets some gas and dust stick out above the average disk height, can be caused by a planetary mass object orbiting in the disk \citep{2011ApJ...736...85U,2013LNP...861..201B}. Alternatively, vortices in the disk can also push material up locally \citep{2010A&A...513A..60L,2012ApJ...754...21L}. 

Changes in the optical brightness and color of a YSO are often attributed to spots, either hot accretion spots or cold magnetic spots, rotating in and out of view or to time-variable extinction \citep{2001AJ....121.3160C,2002AJ....124.1001C}. For two reasons spots are not directly relevant in the mid-IR. First, for all YSOs except class~III sources, the mid-IR flux is dominated by emission from the inner disk \citep{2006ApJS..167..256R} and the stellar photosphere contributes only a small fraction, thus changes in the photospheric flux do relatively little to the observed mid-IR magnitudes. Second, mid-IR wavelengths are already in the long-wavelength limit of the Planck function $B(\lambda,T)$ and thus the resulting variations are nearly colorless, in contrast to the observations in figure~\ref{fig:cmdslope}. The exact $\Delta m$ depends on the temperature $T$ of the photosphere, the hot and cold spot as well as the filling factor $f$ \citep{1986ApJ...306..199V,2001AJ....121.3160C}:
\begin{equation}
\Delta m=-2.5\,\log\left(1-f(1-\frac{B(\lambda, T_{spot})}{B(\lambda, T_{*})}) \right)
\end{equation}
As an example we calculate the change in magnitude and color for a star with $T_{*}=4000$~K.
Assuming a cool spot with $T_{eff} = 3000$~K and a filling factor of $f=10$\% \citep{1989A&A...211...99B,2008A&A...479..827G},  the change in luminosity is $0.04$~mag and the change in color $0.003$~mag. For most stars of our sample fluctuations this small would be masked by the observational noise. The situation is slightly better for hot spots. Assuming a filling factor of 10\% and a hot spot temperature of $20\;000$~K \citep{calvetgullbring}, the change in [3.6] is 0.6~mag, while the [3.6]-[4.5] color changes by 0.05~mag. Again, this is almost colorless and no star in figure~\ref{fig:cmdslope} can be explained by this mechanism alone. If the reddening increases and at the same time the visible fraction of the hot spot decreases, then the resulting slope in the CMD would be intermediate between ISM reddening and colorless. Figure~\ref{fig:cmdslope} shows mostly class~I and flat-spectrum sources in this region of the CMD slope diagram. We conclude that spot modulation does not contribute significantly to the mid-IR variability. Ten of of the 52 cluster members are periodic; in total there are 14 periodic sources out of 61 sources that are variable in the first season. In comparison \citet{2001AJ....121.3160C} find 18\% in $JHK$ monitoring of the Orion~A molecular cloud with a similar time coverage as we have for L1688 and \citet{2013arXiv1309.5300P} find a third of all variable stars in L1688 to be periodic (due to their longer time baseline, \citet{2013arXiv1309.5300P} are sensitive to different periods, but only 4 out of their 32 periods have values that are outside of the range we could detect -- see their table. We are sensitive to periods of up to 15~days here.).

Since spots on the star cannot explain sources that become bluer as they dim, this phenomenon must be related to the disk. Several parameters of the inner disk could be time variable. The most obvious one is the accretion rate. However, the direct effect of increased accretion is a larger accretion spot, which would make the star bluer and brighter, not fainter. Also, \citet{2012PASP..124.1137F} find no relation between mid-IR lightcurves and spectroscopic accretion tracers. This indicates that the relevant parameter can change without effecting the accretion rate. In the mid-IR we see the optically thick dust at the inner disk edge. Its temperature is determined by the distance to the central star and its chemical composition. The distance in turn is mostly given by the stellar irradiation and the dust sublimation temperature. Dust that moves inward of the dust sublimation radius is heated and evaporates to gas on time scales below one orbital period \citep[see discussion in][]{2013AJ....145...66F}. If the irradiation is reduced, for example because the accretion rate drops and the accretion spot cools down, then the dust is reformed and the inner disk radius moves inward. Because the disk is heated over its entire vertical height at the inner edge while only the disk surface absorbs radiation at larger radii, it can form a puffed-up inner rim that reaches above the usual disk scale height and casts a shadow on the remaining disk reducing the luminosity at longer wavelengths. If this rim grows, it intercepts more stellar light. Thus the emission from the inner wall increases and the emission from the outer disk decreases, which results in bluer colors. Since the total emitting area also decreases, the source can become bluer and fainter. However, since the emission at $3.6\mu$m and $4.5\mu$m is formed close to the sublimation radius, this requires changes at those radii, maybe in the shape and not only the height of the inner wall. Time variable self-shadowing has been observed with direct imaging in \object{HD 163296} \citep{2008ApJ...682..548W} and \citet{2010ApJ...717..441E} present a series of \emph{Spitzer} spectra of T~Tauri stars where the spectrum can be modeled taking into account the shadows cast by the inner disk rim. In a time series of spectra those models explain a decreasing far-IR ($>20\mu$m) flux that coincides with an increase in emission from the inner wall \citep{2011ApJ...728...49E}. 

Alternatively, \citet{2013AJ....145...66F} suggested a non-axisymmetric model. If the emission from the stellar photosphere is stronger at some longitudes due to a strong accretion spot or magnetic activity on the star, it irradiates the inner disk like a searchlight beam and causes the dust at this latitude to retreat. The result is a disk where the inner rim is broken at one (or several) positions. This would reduce the emission in $3.6\mu$m, but cause more energy to be emitted at a larger radius, where the dust is cooler on average and thus emits at longer wavelength such as $4.5\mu$m. In this way, the source can become redder as it brightens and bluer when the searchlight beam stops and the dust at the inner rim reforms at all longitudes. In this scenario the time variability of the mid-IR emission should follow the pattern of the searchlight and, if this is emitted by the accretion spot we expect to find the timescale of the stellar rotation period (if the change in brightness is a geometrical effect with the line-of-sight) or a close relation with the accretion rate. With few exceptions, we see the lightcurves changing non-periodically with time scales of just a few days, faster than typical rotation periods of YSOs \citep{1989A&A...211...99B,2008A&A...479..827G}. As for the second idea, this would again predict a relation between the mid-IR luminosity and the accretion rate, which has not been observed yet.

None of the simple models presented so far offers a convincing explanation of the observed lightcurves and CMDs. However, detailed simulations that take into account turbulent transport processes in the disk and different dust species with a complete chemical network have not been performed yet. The real situation might be much more complex than sketched above. Since different dust species form and sublimate with different speeds and at different temperatures, time variable irradiation can actually cause a very complex mixture of species with different opacities, that will not react linearly to increased irradiation. Such a complex network might cause a delayed and non-linear response of the disk, which masks the underlying relation between accretion and disk emission.
