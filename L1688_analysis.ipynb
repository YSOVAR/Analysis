```python
# Fraction of variable sources
from scipy.optimize import brentq
    
def err_bar(k, N, significance = 0.68):
    '''Estimate uncertainty on probability p=k/N, given k and N

    Given an observed number of sucesses k in N trials,
    what is the probability p for sucess?
    Obviously, p=k/N is a good guess. So, what is the
    uncertainty of that guess?
    For a known p, the probability for observe k=k(N,p) is a 
    binomal distribution. To turn this question around, we have to ask:
    How likely is a binomal distribution with probability p to produce
    the observed outcome k,N? 
    This function numerically approximates that
    distribution (called ``fx`` in the function).
    While it is useful to look at the full distribution of p's,
    that is hard to show in plot. To distill it into a simple number
    this function determines the boundaries p0,p1, such that
    fx(p0) = fx(p1) and $\int_p0^p1 fx(p) dp = significance$.

    Parameters
    ----------
    k : int
        number of sucesses
    N : int
        number of trials
    significance : float
        significance interval

    Returns
    -------
    p0, p1 : float
        boundaries of probability p, such that
        fx(p0) = fx(p1) and $\int_p0^p1 fx(p) dp = 0.68$.
    '''
    if not ((significance >=0.) and (significance<=1.)):
        raise ValueError('Significance must be in interval 0..1')
    x = np.linspace(0,1,1000)
    fx = x**k * (1.-x)**(N-k)
    fx = fx/fx.sum()
    pcrit = brentq(lambda y: (fx[fx >y]).sum()-significance, 0., np.max(fx))
    p0 = np.min(x[fx > pcrit])
    p1 = np.max(x[fx > pcrit])
    return np.array([p0,p1])-float(k)/float(N)
```

```python
def robscat(filepath):
    ''' Import Robs Spitzer data

    read Rob's IDL format and make it into a a catalog, deleting multiple columns and add the identifiers
    '''
    s = scipy.io.idl.readsav(os.path.join(filepath, 'cg_merged_srclist_mips.sav'))
    coo=np.ma.array(s.out[:,0:20],mask=(s.out[:,0:20] == 0.))
    s.out[:,20:30][np.where(s.out[:,20:30] < -99)] = np.nan
    s.out[:,30:40][np.where(s.out[:,30:40]==10)] = np.nan
    
    dat = Table()
    dat.add_column(Column(name='RA', data=np.ma.mean(coo[:,[0,2,4,12,14,16,18]],axis=1), units = 'deg', format = '9.6g'))   #RA is avarage of all valid (non-zero) Ra values in 2MASS JHK, IRAC 1234
    dat.add_column(Column(name='DEC', data=np.ma.mean(coo[:,[1,3,5,13,15,17,19]],axis=1), units = 'deg', format = '+9.6g'))

    robsyukyformat={'J_MAG': 20,'H_MAG': 21, 'K_MAG': 22,'J_ERR': 30,'H_ERR': 31,'K_ERR': 32,'IRAC_1': 26,'IRAC_2': 27, 'IRAC_3': 28, 'IRAC_4': 29,'IRAC_1_ERR':36,'IRAC_2_ERR':37, 'IRAC_3_ERR':38, 'IRAC_4_ERR':39}
    for col in robsyukyformat.keys():
        dat.add_column(Column(name=col,data=s.out[:,robsyukyformat[col]], units = 'mag', format = '4.2g'))

    s.mips[:,2][np.where(s.mips[:,2] == -100)] = np.nan
    s.mips[:,3][np.where(s.mips[:,3] == 10)] = np.nan
    dat.add_column(Column(name='MIPS', data=s.mips[:,2], units = 'mag', format = '4.2g'))
    dat.add_column(Column(name='MIPS_ERR',data=s.mips[:,3], units = 'mag', format = '4.2g'))

    IRclass = scipy.io.idl.readsav(os.path.join(filepath, 'cg_classified.sav'))
    dat.add_column(Column(name='IRclass', dtype='|S5', length=len(dat)))
    for n1, n2 in zip(['wdeep', 'w1','w2','wtd','w3'], ['I*', 'I', 'II', 'II*', 'III']):
        if n1 in IRclass.keys():
            dat['IRclass'][IRclass[n1]] = n2
    dat.add_column(Column(name='AK', data=IRclass.ak, units = 'mag', format = '4.2g'))
   
    return dat

rob = robscat(inroot + '/rob09/')
rob.add_column(Column(name='robsave_id', data=np.arange(len(rob))))
names = [['IRAC_1','3.6mag'], ['IRAC_2','4.5mag'], ['IRAC_3','5.8mag'], ['IRAC_4','8.0mag'], ['MIPS', '24mag'], ['J_MAG', 'Jmag'], ['H_MAG', 'Hmag'], ['K_MAG', 'Kmag'], ['IRAC_1_ERR','e_3.6mag'], ['IRAC_2_ERR','e_4.5mag'], ['IRAC_3_ERR','e_5.8mag'], ['IRAC_4_ERR','e_8.0mag'], ['MIPS_ERR', 'e_24mag'], ['J_ERR', 'e_Jmag'], ['H_ERR', 'e_Hmag'], ['K_ERR', 'e_Kmag']]
for n in names:
    rob.rename_column(n[0], n[1])
for n in ['3.6mag','4.5mag','5.8mag','8.0mag','24mag']:
    rob.add_column(Column(name=n+'_s', length=len(rob), dtype='S5'))
    rob[n+'_s'][np.isfinite(rob[n])] = 'G09'
for n in ['Jmag','Hmag','Kmag']:
    rob.add_column(Column(name=n+'_s', length=len(rob), dtype='S5'))
    rob[n+'_s'][np.isfinite(rob[n])] = '2MASS'
    
robnames= set(rob.colnames) - set(('RA', 'DEC'))

cat.add_catalog_data(rob, names = robnames, ra1='ra', dec1='dec', ra2='RA', dec2= 'DEC')
```

If you dig into one of Maria's papers or the original YSOVAR proposal, we chose a repeating 3.5 day cadence with time-steps of 2,4,6,8,10,12,14,16,12 hours with +/- 1 or 2 hour tolerances.

The +/-1 or 2 hour tolerances introduced a very useful randomization of the cadence.

The time-steps were chosen to sample different frequencies evenly (ie, d delta_t /dt = constant).  The smaller time-steps made us sensitive to variability as short as 2 hours (high frequencies; and in the case of Orion it was 4 hours, since it took 4 hours to do a single map), without increasing our overall observing program to many more hours than it was actually executed under, or had much smaller time-baselines.  Similarly the longer time-steps made us sensitive to low frequency variations.

The two items above served a very important purpose - avoid 1-day aliasing.  We were in space and there's no reason to have any cadence aliasing.

The 3.5 day repeating pattern was chosen to fit two cycles every week, which is how Spitzer did it's scheduling at the time - in one week chunks.

```python
plt.plot(cat['coherence_time_36'], cat['coherence_time_45'],'rs', label = 'Coherence time')
plt.plot(cat['cumsumtime_36'], cat['cumsumtime_45'],'bo', label = 'Cumulative sum')
plt.xlabel('IRAC1')
plt.ylabel('IRAC2')
plt.title('Compare different measures of the ACF')
plt.xlim(0,14)
plt.ylim(0,14)
plt.legend()
```

The plot above compares the measures of the autocorrelation function (ACF) for IRAC1 and IRAC2. If both bands are correlated (some objects might have strong color variations, but that is the exception), then the ACF for both bands should be similar.

The plot compares the coherence time (the width of the first peak in the re-sampled ACF) and the cumulative sum (a measure with a similar idea, but defined for the non-resampled ACF) for IRAC 1 and IRAC 2. For short times (t < 2. days) it's not quite clear that they are related, but for longer times the number obtained from both channels seem to agree to some extend, indicating that the values measured are real and not just noise.

Another way to plot this is to compare the coherence time and the cumulative sum for a single channel and then see if the statistical distributions are similar in both channels.

```python
plt.plot(cat['coherence_time_36'], cat['cumsumtime_36'],'rs', label = 'IRAC 1')
plt.plot(cat['coherence_time_45'], cat['cumsumtime_45'],'bo', label = 'IRAC 2')
plt.xlabel('coherence time')
plt.ylabel('cumsum time')
plt.title('Compare different measures of the ACF')
plt.xlim(0,10)
plt.ylim(0,14)
plt.legend()
```

Ignoring a few outliers with long coherence time and also long cumsum time, the distribution seems anti-correlated as expected. The outlier might be lightcurves with a low signal. We can test that be selecting them and looking at their lightcurves or ACFs visually. Below, I inspect the most extreme examples.

```python
ind = (cat['coherence_time_45'] > 4.) & (cat['cumsumtime_45'] > 4.) & np.isfinite(cat['cumsumtime_45'])
for i in ind.nonzero()[0]:
    t = cat.lclist[i]['t45']
    print '{0} datapoints in lightcurve {1}'.format(len(t[t<55340]), i)
    plt.plot(t, cat.lclist[i]['m45'])
plt.xlim(55290,55340)
ax = plt.gca()
ax.invert_yaxis()
ax.set_xlabel('MJD')
ax.set_ylabel('[4.5]')
```

All those lightcurves are fairly faint and fairly noisy. Also, some of them have long gaps in the data. When building up the ACF these gaps get interpolated linearly, which will cause a long coherence time in the resulting ACF. The cumsumtime uses the original data without any resampling, hence it does not suffer from this problem. It's long because there is no apparent short-term correlation, but also because there are very few data pairs with a small $\delta t$.

So, I conclude that the measures of the non-periodic time scale from the ACF are useful, if either of the time scales is > 2 days (in which case the other one will be short, since they are anti-proportional).

## Introduction

Our knowlegde on star formation has developed tremendously in the last few years, not the least due to new observational oppotunities on ground- and space-based infrared observatories. Stars form in dense and cool molecular clouds. When the density is high enough in a region, the matter can graviationally collaps and form a young stellar object (YSO). In the early phases, this infall happens almost radially, and the thick envelope dominates the emission from the object and hides the details of what is going on within (class~I). In the IR the dust opacity is much less severe than in the optical and thus observations at longer wavelengths are particularly suited to observe those objects, many of which were discovered in \emph{Spitzer} surveys in the $24\mu$ channel. Due to the conservation of angular momentum, the surrounding envelope eventually flattens out to form a circum-stellar disk. If the inclination of the system is not edge-on, then the photospheric emission from the central YSO can be observed. However, the disk still causes an IR excess above the level of the stellar photosphere (class~II), which can be used to distinguish those objects from more-evolved stars, e.g. using the \emph{Spitzer} IRAC camera. The accretion disk does not reach down to the central star, instead its inner radius is given by the dus-sublimations temperature. Some of the mass in the circum-stellar disk is accreted onto the central star via magnetically confined accretion funnels, some is blown out in disk winds and some condenses into planets. When the disk is cleared out, the IR colors cannot be distinguished from main-sequence stars any more (class~III).

In addtion to the circum-stellar absorption many YSOs are still surrounded by their molecular cloud, with can add extra dust extinction, so that even more evolved objects often appear reddened and are invisible in the optical. The closest optical counterpart of the class~II sources is the class of T~Tauri stars (TTS), which was originally identified by their variability \citep{1945ApJ...102..168J}, long before anybody realized that TTS are indeed pre-main sequence stars. Optical monitoring of CTTS has a long tradition. In most cases, particularly in class~II and III sources, the dominant time scale is the stellar rotation period of the a few days \citep{1983ApJ...267..191R,1986A&A...165..110B,2009ApJ...694L.153N}. YSOs can have both cool spots caused by magnetic activity similar to the spots observed on our Sun and hot spots which mark the impact points of the accretion funnels onto the stellar surface. This impact happens at free-fall velocities up to 500~km~s$^{-1}$, thus the accretion shock heats the accreted mass to X-ray emitting temperatures \citep[e.g. review by][]{2011AN....332..448G}. In the optical, the accretion shock appears as additional emission that can be approximated as a black-body with $10-20\,000$~K (\citealp{calvetgullbring}, but see also \citealp{2012AstL...38..649D,2013AstL...39..389D}). Variability in the mass accretion rate can leads to changes in the hot spot signatures. We expect the rotation period of the inner disk where the accretion funnels start to be the typical time scale for those processes and indeed optical variability with amplitudes around 0.1~mag is often observed in CTTS \citep{1996MNRAS.282..167S,2008MNRAS.391.1913R}. Another source of variability related to the accretion could be shock oszillations. This has been predicted theoretically \citep[e.g.][]{2008MNRAS.388..357K}, but is not observed so far \citep{2010A&A...518A..54G}.

YSOs can also vary on much longer time scales. Variability on the scales of years could be caused by changing circumstellar extinction \citep{2007A&A...461..183G} or by massive accretion events when a significant fraction of the disk mass drains onto the YSO. In this case the accretion luminosity can outshine the YSO by orders of magnitudes and if takes months to years \citep[in the case of EXor outbursts][]{2012ApJ...749..188L} or even centuries \citep[FUor outbursts][]{1996ARA&A..34..207H} until the accretion decays back down to the original level.

The spectral energy distribution (SED) of YSOs in the optical is dominated by the stellar photosphere and the accretion spot. Thus, optical monitoring is very effective to understand the stellar rotation and the accretin spot. The disk radiates more strongly at longer wavelengths. Depending on the mass in the disk it will start to dominate the SED at $K$ band or in the IRAC bands at $3.6\mu$ and $4.5\mu$. Simple disk models still treat the disk as a static and axisymmetric structure, but observationally it now seems that the disk is in fact ``a bubbling, boiling, wrinkled, dented, warped mass of gas and dust'' \citep{2013AJ....145...66F}. The YSOVAR (Young Stellar Object VARiability) project has monitored the \object{Orion nebular cluster} and eleven smaller star forming regions with \emph{Spitzer} in the warm mission. Data from several different observing programs are analyzed in the context of the YSOVAR project. The largest portion of the telescope time was awarded to PI Stauffer as a Cylce 6 Exploration Science program. First results of the YSOVAR project on the ONC are published in \citet{2011ApJ...733...50M} and \citet{2012ApJ...753..149M}. More details of the observing strategy and an overview of the data can be found in \citet{luisa} (from now on ``the YSOVAR paper''). In this article we present results that are specific to the star forming region \object{L1688}.

Lynds~1688 (L1688) is a sub-cloud of the $\rho$~Oph star forming region, one of the best-studied star forming regions in the sky \citep[review by][]{2008hsf2.book..351W}. The distance to the cloud is a matter of active discussion. VLBA measurements and Hipparcos parallaxes point to values around 120~pc \citep{2008ApJ...675L..29L,2008A&A...480..785L}, while \citet{2008A&A...480..785L} find 135~pc, again based on Hipparcos data. One possible solution is that the streamers of L1688 are closer to the Sun than the core of the region \citep{2008A&A...480..785L}. In this article, we adpot a distance of 120~pc, but we note that most of our results are independent of the actual distance. The central region of L1688 is very dense and deeply embedded ($A_V=50-100$~mag), thus no complete survey of all cluster members exisits, but an extinction limited, spectroscopic survey \citep{2011AJ....142..140E} finds an average age of 3.1~Myrs for a 6.8~pc$^2$ region centered on L1688 and no significant deviation from the interstellar mass function. Earlier studies \citep{1995ApJ...450..233G,1999ApJ...525..440L, 2002A&A...393..597N} concentrated on the deeper embedded core and found a much younger age of 0.3~Mys. Some, but not all of this difference is due to the specific reddenign laws or pre-main sequence evolutionary tracks used in these studies \citep[see discussion in][]{2011AJ....142..140E}.

In this article, we present global properties of the variability in L1688. Noteable individual objects are studied separately. \citet{2012PASP..124.1137F} present near-simultaneous spectroscopic monitoring for five YSOs in L1688. They do not see any correlation between the hydrogen emission lines that are accretion indicators and the features in the IR lightcurves of their targets. Noteable objects with well-sampled NIR lightcurves are \object{WL 4} \citep{2008ApJ...684L..37P} and \object{YLW 16A} \citep{2013A&A...554A.110P} which show eclipses with periods of 131 and 93~days, respectively, that can be interpreted multiple systems, where one or more components are ecliped by a warped circumstellar disk.

In section~\ref{sect:observationsdatareductionandauxiliarydata} we introduce the data reduction and discuss sources lists and stellar properties obtained from the literature. Section~\ref{sect:results} contains the observational results such as the characterization of the lightcurves as variable or even periodic, the color changes of sources and how these findings depend on the class of the object. Section~\ref{sect:discussion} discusses physical models to explain the observed features in the lightcurves. We end with a summary and some conclusions in section~\ref{sect:summaryandconclusions}.

Three fields in L1688 were observed with \emph{Spitzer} in four observing periods from 2010-04-12 to 2010-05-16, 2010-09-22 to 2010-10-27, 2011-04-20 to 2011-05-23 and 2011-10-01-2011-11-06. In the first observing period the sampling is much denser. About 70 epochs have been observed on most sources with irregular time intervals to aid period detection and avoid aliasing with the observations frequency. The time between two consecutive epochs in this period is between 3 and 17 hours. In the remaining three observations periods the sampling is much more sparse with a typical interval between epochs of a few days. Less than ten datapoints per source were taken in the later three observation periods. In total there are 108 epochs of observation with total exposure time of 30.7 hours. All epochs can be found under Program Identification umber (PID) 61024 in the Spitzer Heritage Archive. The timing of the observations is constrained by the visibility of the region for the \emph{Spitzer} spacecraft and the requirement to be sensitive to time scales of variability as short as a few hours and as long as two years. Due to a slight field rotation over the observing periods sources on the edge of the field are less well sampled. The three fields chosen were observed with the IRAC1 and the IRAC2 channel (effective wavelength 3.6$\mu$m and 4.5$\mu$m). Both channels operate simultaneously, but their fields-of-view are non-overlapping patch of sky. Thus, the three fields chosen are observed in two consecutive pointings, one for each channel. In each observation of one of the target fields we automatically obtain data from a neighboring fields observed by the channel that is offset from the target field. In spring sources to the south of the main fields only have IRAC1 data and those to the north only have IRAC2 data. In the fall, the situation is reversed. We include sources in these neighboring fields in the analysis, if they are observed more than 5 times in one channel. In each epoch data was taken in IRAC mapping mode using the high-dynamic range (HDR) data acquisition, which obtains an 0.4~s and an 10.4~s exposure for each pointing.

A very detailed account of the data processing and the source extraction for the data is given in the YSOVAR paper and we refer to that publication for details. Here we summarize the main data reduction and processing steps. Basic calibrated data (BCD) are obtained from the \emph{Spitzer} archive. Further data reduction is performed with the IDL package cluster grinder \citep{2009ApJS..184...18G}, that treats each BCD image for bright source artifacts, and combines images into mosaics, rejecting cosmic rays in the process. For each epoch and each channel, aperture photometry is run in individual BCDs with an aperture radius of $2\farcs4$. The photometry reported for an epoch is the average brightness of all BCDs within that epoch that contain the source in question, after rejecting outliers (e.g.\ due to cosmic rays). The photometric uncertainties obtained from the aperture photometry are, particularly for faint source, only lower limits to the total uncertainty, since distributed nebulosity often found in star forming regions can contribute to the noise. To improve these estimates the YSOVAR paper introduces and error floor value that is added in quadrature to the uncertainties of individual photometric points. The value of the error floor is 0.01~mag for IRAC1 and 0.007~mag for IRAC2.

As described in the YSOVAR paper, we cross-match sources from individual observations with a matching radius of $1\arcsec$ with each other and with the 2MASS catalog, which is used as a coordinate reference. All photometric measurements performed in the context of the YSOVAR project are collected in a central database, which we intend to deliver to the IRSA for general distribution.

### Auxiliary data from the literature

The star forming region L1688 has been the target of intense study over the past decades and a wealth of additional information exists in the literature. In particular, we refer the reader to two reviews \citep{1992lmsf.book..159W} and \citep{2008hsf2.book..351W}. The latter review compiles a list of objects with a high probability of membership from a variety of published sources. The membership criteria employed are detected X-ray emission, which is only detectable at young, active stars at the distance of L1688, optical spectroscopy, with H$\alpha$ in emission, Li in absorption or a location above the main-sequence in the HR diagram, as well as IR emission that is indicative of the a circum-stellar disk as expected in YSOs. We use the membership table of \citet{2008hsf2.book..351W} to establish membership in the L1688 star forming regions for the sources observed. More recently, the regions was surveyed by \citet{2012ApJ...751...22B} in $JHK$, to a level significantly deeper than in 2MASS. This leads to a significantly increased number of members in L1688. Due to the lower photometric uncertainties, we use the $JHK$ magnitudes from \citet{2012ApJ...751...22B} where available.

L1688 was also observed with \emph{Spitzer} in the initial cold mission phase with all four IRAC channels and the 24$\mu$m channel of the MIPS. Objects classified as YSOs from these data are already contained in the membership list of \citet{2008hsf2.book..351W} but we note that this data has been reprocessed for consistency with our new observations using the data reduction process for \emph{Spitzer} data described above. We cross-match our detections with that catalog using a matching radius of 1\arcsec and if we did not obtain a photometric value for an IRAC band or the $24\mu$ MIPS, but a value with the quality specifier A, B or C is present in the c2d catalog then we use that value. The source tables contain a column per band that specifies if a data point is taken from our own data reduction \citep[``gut''][]{2009ApJS..184...18G} or the c2d database \citep[``c2d''][]{2003PASP..115..965E}.

The YSOVAR data is cross-matched with the SIMBAD catalog, the NOMAD catalog \citet{2004AAS...205.4815Z}, the \emph{Spitzer} cold mission source list from \citet{2009ApJS..184...18G}, which include a classification of objects based on a series of cuts in color-color and color-magnitude diagrams, which is more stringent that the classification we employ here and the lists of cluster members from \citet{2008hsf2.book..351W} and \citet{2012ApJ...751...22B}. ### Instrumental effects on IRAC lightcurves

```python
def characterize_seasonal_jump(t, m, merror):
    '''Characterize the seasonal jump (an instrumental effect)

    This is +1, is the mean in season 1 and 3 is higher than the mean in season 2 and 4,
    -1 if it is vice-versa and 0 if the mean values for each season do not correlate with the 
    time of year the obervations were taken.
    ``np.nan`` is returned for lightcurves that do not have data in all seasons.
    '''
    seasons = np.array([[55295,55338], [55458,55500], [55668, 55710], [55830, 55890]])
    mean = np.zeros(4)
    n = np.zeros(4)
    for i in range(4):
        ind = (t > seasons[i,0]) & (t < seasons[i,1])
        mean[i] = np.mean(m[ind])
    ind = np.argsort(mean)
    if (set(ind[0:2]) == set([0,2])) & (np.min(mean) > 0):   # the mean > 0 makes sure that data exisits in all seasons.
        return 1
    elif (set(ind[0:2]) == set([1,3])) & (np.min(mean) > 0): 
        return -1
    elif (np.min(mean) > 0):
        return 0
    else:
        return np.nan
    
YSOVAR.registry.register(characterize_seasonal_jump, error = True, time=True, force=True, name='seasonaljump')
```

Despite the careful data reduction described above, some residual artifacts remain in the \emph{Spitzer} lightcurves. L1688 is particularly suited to find those since most instrumental artifacts are closely related to the position on the detector. The three target fields observed in L1688 have almost no rotation within one season so that each source falls onto the same region (albeit a different pixel) on the chip. The spacecraft orientation flips between seasons, so that most intrumental artifacts will lead to lightcurves that have one level in season one and three and another in season two and four. Figure~\ref{fig:instrumentallcs} shows examples of lightcurves that might fall into this category.

some sources.\label{fig:instrumentallcs}}
\end{figure}
Intrinsic variability can mask this effect if it is lage enough or it can mimic this effect if the source is intrinsically variable with a period of one year. Thus, we visually inspected every frame for a sample of sources with lightcurves similar to those in the figure. Different sources are influenced by different instrunemtal effects. Some are close to a detector edge, so that the background is not well-determined, some are in the wings of the point-spread function (PSF) of a bright neighbor and some show residuals from hot pixels. There are other sources which appear perfectly normal when looking at the frames. In those sources the variability could eiter be real or they might fall in a region where the large scale gain-map correction is not accurately calibrated. Most of these effects would only affect one of the spacecraft orientations; even the PSF wings change with the orientation since the PSF is not perfectly circular. Figure~\ref{fig:instrumentallcs} shows sources with a magnitude between 8 and 15 in $4.5\mu$. The relative difference in magnitude is smaller for brigther sources, but since the photometric errors also are smaller these instrumental effects can still be significant. 

In order to quantify the number of lightcurves that suffer from this problem we use a statistical approach. For each channel we have about 200 lightcurves with datapoints in all four observing seasons. We calculate the mean magnitude in each season and test weather the two brightest mean magnitudes are either from season one and three or season two and four. If the source is constant and the difference in magnitude is only obervational noise, we expect that $\frac{1}{6}$ of all lightcurves have the brightest mean magnitudes in season one and three, in $\frac{2}{3}$ the brightness of the magnitudes is not correlated with the observing season and $\frac{1}{6}$ are brightest in season two and four. Given the large number of sources with different intrinsic time scales even sources with intrinsic variability should have the distribution of $\frac{1}{6}:\frac{2}{3}:\frac{1}{6}$. In contrast, the observed lightcurves have the distribution of 49:79:70 for IRAC1 and 57:103:68 for IRAC2. This is incompatible with the expected multinomial distribution (p-values $<10^{-10}$). Comparing the expected and the actual distributions, we see that in each band 50 lightcurves, a quarter of the sample, are missing from the bin where the brightness is not correlated with the spacecraft orientation. This implies that about a quarter of all lightcurves suffers from the artifacts described above. It is almost equally likely to find the maximum brightness in season one and three or two and four. However, most of the artifacts depend on the specific detector and thus the probability to find correlated variability in both channels is only about $0.25^2/2=3$\%. 

In section~\ref{sect:variability} we determine the variable members using a Stetson and a $\chi^2$ test as defined in the YSOVAR paper. The limits in those tests are designed to be conservative and indeed we find a distribtion of 6:27:11 for IRAC1 and 5:32:8 for IRAC 2, when we restrict the sample to those lightcurves that will be classified as variable below. Both ratios are fully compatible with the expected multinomial distribution (p-value 0.3 and 0.7, respectively), which shows that the limits we apply are conservative enough that the sample of variable stars only contains sources where the variability is totally dominated by the intrinsic source variability.

The other star forming regions observed by the YSOVAR project have a higher field rotation within one season, so there is no discrete switching point and the instrumental artifacts discussed here will not result in a step in the lightcurve. This makes them more degenerate with intrinsic stellar variability, but given that the observations all use the same setup and the same definition for intrinsic variability as we do here, we expect that the sources classified as variable in other YSOVAR studies also have a negligible contribution from the instrumental artifacts discussed in this section.

Only 77 of all 984 sources are members of the L1688 star forming regions according to \citet{2008hsf2.book..351W}. Just over 10\% of the 984 sources are classified as stars by \citet{2009ApJS..184...18G}, the remaining sources are either rejected as contaminants such as AGN, star forming galaxies or foreground stars that are blended with the nebulosity of the cloud or they cannot be classified by their algorithm because the source is detected in few bands only. Of those sources, 17 are class~I sources, 36 are class~II sources and 59 are class~III, i.e.\ disk-less sources, which can either be pre-main sequence stars that lost their disks only recently, or main-sequence stars. The \citet{2009ApJS..184...18G} classification scheme is very conservative, it represents a set of sources that can be dereddened and classified to a high degree of confidence. However, for many sources we lack the required coverage in many bands to perform this classification. Thus, the YSOVAR paper defines a simpler approach, which uses the observed colors only. In this scheme, we fit the spectral slope $\alpha=\frac{d \log \lambda F_{\lambda}}{d \log \lambda}$, where $\lambda$ is the wavelength and $F_{\lambda}$ the flux per unit wavelength interval at that wavelength. We make use of all measured fluxes (no upper limits) in the 2-24$\mu$m range, which corresponds to the range from the $K$ filter to the MIPS 24$\mu$m channel. We use the fluxes for all *Spitzer* bands from the \citet{2009ApJS..184...18G} paper and summarize the new lightcurves by calculating the mean and standard deviation for the new IRAC1 and IRAC2 data. Thus, sources detected both in the cold mission and in the new dataset, will have two datapoints for IRAC1 and IRAC2. We perform a least-squares fit to the available data and call sources with $0.3 < \alpha$ class~I, $-0.3 < \alpha <0.3$ flat-spectrum (F), $-1.6<\alpha < -0.3$ class~II and $\alpha < -1.6$ class~III sources. In this scheme a significantly reddened main-sequence star, that has an intrinsic slope $<-1.6$ may appear as an class~II, a flat-spectrum or even class~I object, so this observational classification cannot be translated directly into the evolutionary stage of an object. However, only a detection in any two bands is required to calculate $\alpha$, thus we can assign a classification to 895 of the 984 objects. We find 108 class~I, 58 flat-spectrum, 285 class~II and 444 class~III sources. Comparing those sources, which are also identified in \citet{2009ApJS..184...18G}, the resulting classification is very similar, particularly for the class~I and flat-spectrum sources. All sources with a spectral slope $<0.3$ are also class~I sources in that paper, with two exceptions our flat-spectrum sources are either class~I or II in \citet{2009ApJS..184...18G}. We use three different methods to detect variability in the lightcurves. For details on those methods see the YSOVAR paper. Sources in the targeted fields are observed in in IRAC1 and IRAC2 within a few minutes. For those sources we calculate the Stetson index $s$ with points weighted evenly \citep{1996PASP..108..851S}.
\begin{equation}
s = \frac{1}{\sqrt{N(N-1)}}\sum_{k=1}^{N}sig(P_k)\sqrt{|P_k|}
\label{eqn:stetson}
\end{equation}
where the sum is taken of all $N$ pairs of observations in IRAC1 and IRAC2 with observed magnitudes $a_k$ and $b_k$ and uncertainties $\sigma_{a_k}$ and $\sigma_{b_k}$. $sig$ denotes the sign of $P_k$. $P_k$ is the product of the normalized residuals in both bands:
\begin{equation}
P_k  = (\frac{a_k-\bar{a}}{\sigma_{a_k}}) \times (\frac{b_k-\bar{b}}{\sigma_{b_k}})
\end{equation}
Here $\bar{a}$ denotes the error weighted mean of all IRAC1 magnitudes and $\bar{b}$ the error-weighted mean of all IRAC2 magnitudes. The Stetson index is very robust to observational errors since those are unlikely to effect both bands in the same way. We find 47 sources to be variable, that is 20\% of all lightcurves where we can calculate the Stetson index. The number is slightly lower (40 sources, 18\%) is we consider only the first observing season. This is not surprising, given that the Stetson index is designed to be a "robust" statistic, that is little influenced by a small number od datapoints and by far most datapoints belong to the first season. We caution that equation~\ref{eqn:stetson} will introduce a systematic bias in such a way that a source that is variable in ony one band will not be marked as variable. This is necessary, because about an eigth of all observed lightcurves show seasonal variations in the one channel, such that the magnitude is always higher in spring observing epochs than in fall epochs or vice-versa due to the orietation of the space craft.

Figure \ref{fig:stetsonlc} shows two examples of lightcurves that are classified as variable according to the Stetson index.

For sources outside the primary target fields the Stetson index cannot be calculated, since lightcurves are observed in one band only. For those sources we calculate the reduced $\chi^2_{red}$ to the mean of the lightcurve. Instrumental uncertainties lead to a non-Gaussian error distribution and and we use a very conservative cut-off and mark sources as variable only if $\chi^2_{red} > 5$ (YSOVAR paper). Using this metric, we find 49 variable sources in the IRAC1 and 38 in IRAC2. If --instead of comparing to the mean-- we fit a line, only one source is classified differently. Given the seasonal instrumental effect on the $4.5\mu$ flux mentioned above, sources where the IRAC2 lightsurve is constant within each season and varies only with the space craft orientation would likely be spurious.
When fitting higher order polynomials, the number of sources with $\chi^2_{red} > 5$ goes down, e.g. for a quadratic polynomial only 41 and 30 sources have a $\chi^2_{red}> 5 $ in IRAC1 and IRAC2, respectively (figure~\ref{fig:chi2lc}). 
A second way to express this is to calculate the $\chi^2_{red}$ only for the first season of monitoring for those sources that are detected in more than one season. Many of those have significantly larger values of $\chi^2_{red}$ for the whole observation series than within one season indicating a constant luminosity over 40~days, but change it over time scales between six months and two years. Contrary to the Stetson index, the $\chi^2_{red}$ is more sensitive to a few outlying data points and thus is easily detects long-term variability, which might not be found by the Stetson index due to the small number of measurements in the later seasons.

In principle the SED classes of YSOs, class~I, flat-spectrum, II and III represent an evolutionary sequence. Initially, the entire emission from the YSO is dominated by the dusty envelope (class~I). As the system evolves the dust settles in a disk (class~II) until this disk disperses entirely due to planet formation, accretion and evaporation and all IR emission is photospheric. YSOs in this stage cannot be distinguished from main-sequence objects by their IR emission alone (class~III). In practice, the observational SED scale that we use here will not follow this path exactly. One complication is that intermediate dust, that is not associated with the specific YSO, also reddens the spectrum and makes the star appear younger than it actually is, e.g. background star that happens to be found behind a this layer of the cloud might appear as class~II or even F star. Also, the orientation of the specific system plays in important role. Already at very early stages (class~I) YSOs can develop powerful outflows, which pierce their envelope and allow us to view the star through the outflow cavity, so that the photospheric contribution becomes important in an early stage. On the other hand, if the see the accretion disk edge-on it will cause a high degree of reddening even for late evolutionary stages, when the envelope is already dispersed. Simulations show that stars of very different ages can thus appears with a similar SED \citep{2006ApJS..167..256R}. These effects tend to smooth out age effects when comparing different evolutionary classes, as each bin will contain a few sources of neighboring evolutionary stages.

Similarly, the membership lists for L1688 might not be complete, so that some sources that we treat as non-members here are actually YSOs, or contain erroneously included stars, which have a YSO-like SED do to reddening by the intervening cloud.
Figure~\ref{fig:probvarSED} shows the fraction of variable stars with a mean magnitude $<14$ for cluster members and non-members. For fainter stars the variability cannot be reliably detected, unless the amplitude is exceptionally large. The observed fraction of variable stars is the best-guess for the probability to find variability for a star of this evolutionary class. However, some bins in the histogram contain only a small number of sources, so this probability is not very well constrained. We calculate the uncertainty for the probability based on the observed number of variable stars $v$ in a bin with $N$ total sources as follows: If the true probability for a sources in the bin to be variable is $p$ then the probability $p_{obs}$ to observe $v$ variables in $N$ stars is given by a binomial distribution:
\begin{equation}
p_{obs}(v,N,p) = {v \choose N} p^v (1-p)^{N-v}
\end{equation}
The peak of this distribution is $p_{obs}(p) = \frac{v}{N}$. We then calculate the boundaries of the confidence interval $p_1 .. p_2$ for $p$ such that $p_{obs}(p_1) = p_{obs}(p_2)$ and the area under the curve includes 68\% of the probability. Those uncertainty ranges are shown in figure~\ref{fig:probvarSED}. The observed probability for all classes of non-members is compatible with a variability fraction just under 10\%. The fact that this barely changes with the evolutionary class, much unlike the distribution for the cluster members, indicates that the contamination with unidentified cluster members is low since they would bias the observed variability fraction to higher values for the younger classes and the vastly different number of sources in the different bins make it highly unlikely, that each bin is contaminated by the same fraction of class~I-II cluster members.

All known cluster members in class~I are variable. This fraction decreases to about 70\% in class~II sources and then drops rapidly. In class~III sources it is already compatible with the fraction observed for non-members. This distribution is a powerful indication that the main cause for the variability must be found in the inner disk regions. In class~I and F sources a significant fraction of the observed IR radiation comes from the envelope, but the variability is still present when the envelope disperses and the mass concentrated in the disk (class~II). However, at the point when the inner disk vanished, the variability suddenly becomes undetectable.

Figure~\ref{fig:meanampSED} looks in more detail at the variable sources using the $4.5\mu$m data. The plot for $3.6\mu$m looks similar, but is not shown here. The figure shows the mean amplitude of the observed variability, where we define the amplitude as the difference between the 10\% and the 90\% quantile of the observed lightcurve. The errorbars in this figure represent the standard deviation of the mean amplitudes within one class. This time there seem to be some differences for the non-members, but we note that the values for non-member flat-spectrum sources are based on a very small sample. Thus, only one or two unclassified cluster members should already explain this difference. However, for the members we can clearly see that the variability is stronger in class~I and flat-spectrum sources than in class~II sources, where the amplitude is already compatible with class~III sources. Specifically, the spread of the earlier classes is much larger than for the later classes.

In summary we can say that the probability for a source to be observed as variable decreases only little between class~I and II sources, but the mean amplitude as well as the differences within a class are much larger for younger sources which have more circumstellar material.

As the last aspect in this section we will look at the timing behaviour of the lightcurves. The list of periodic sources contains objects of all evolutionary stages, but the number is too small to recognize any trends with the evolutionary stage. Instead, we calculate the auto-correlation function $ACF(\tau)$ for each lightcurve. By definition $ACF(\tau=0)=1$ and the $ACF$ decays for longer time-lags $\tau$. We take the first value of $\tau$ with $ACF(\tau)=0.5$ as the characteristic time-scale for a lightcurve. A more common definiton would be to use the position of the first local maximum in the $ACF$, but due to the low number of datapoints in our lightcurves, the noise in the auto-correlation function is large and this value is often not very well defined. While our definition might not give us the time scale of the physical processes at work, they are still valid for relative comparisons. One complication here is that our data is unevenly sampled. In order to calculate the $ACF$ we linearly interpolate the lightcurve on a grid with time steps of 0.1~days. This process can clearly change the properties of teh lightcurves on times scales shorter than the distance between two observations (0.1 to 0.8 days). However, for most lightcurves the relevant time scales are longer than this. We find that the average time scale decreases from 4~days for class I sources to 1.5 days for class~III sources. A more detailed analysis of stochastic and quasi-periodic properties of the lightcurves is beyond the scope of the current paper.

For those sources we compare the color and the magnitude in the $3.6\mu$m band. Figure~\ref{fig:cmd} shows the color-magnitude diagram (CMD) for four sources. In some cases, the color is fairly stable within one season, but changes between seasons, in others the time scale of variability is much shorter and color changes are seen within one season as well in just the same way as different objects show different timescales for variability in the single band lightcurves. For non-variable sources the CMD forms a point cloud, but for variable sources the shape of the CMD can reveal the physical cause of the variability, e.g.\ if a disk warp for accretion funnel passes in front of a YSO we expect its color to become redder as it becomes fainter. If the absorber has the same gas and dust properties as ISM, then the datapoints in the CMD would follow line with the slope of the interstellar reddening law. Based on those consideration we fit a straight line to all points in the CMD using an orthogonal distance regression method that takes the errors in x and y direction into account. From a total of 47 variable sources with CMDs with a sufficient number of datapoints 39 can be fitted with a formal fit uncertainty on the line slope below $6^{\circ}$. The CMD of the remaining sources cannot be descibed by a straight line. Most of those take the form of an uncorrelated scatter plot, either because the source is very faint and dominated by noise or several physical processes that lead to different slopes are active at the same time. As can be seen in figure~\ref{fig:cmd} even in those sources with a well defined slope in the CMD the scatter around the best linear fit is significantly larger than expected from the measurement uncertainty.

There are a few sources where the statistical error on the slope in the CMD is large. Most of those are either faint and thus have large observational error on each individual datapoint or they only show very little variability, so that a possible slope in the CMD is hidden in the intrinsic scatter that is observed in each of the CMDs (figure~\ref{fig:cmd}). Source 174 is an exception and its lightcurve and CMD for the first season of observations are shown in figure~\ref{fig:cmdsquare}. It has a class~II SED slope and is a YSO of low mass \citep[spectral type M1][]{1999ApJ...525..440L}. There is no single slope in the CMD, instead the sources seems to switch between different modes. Initially, the luminosity drops sharply, while the color is almost unchanged or becomes even slightly bluer. The behavior changes around MJD 55305. For the next 15 days, the slope in the CMD is a little less steep than the ISM reddening law. The evolution is not monotonous, but a brightening (e.g.\ around MJD 55313) corresponds to a bluer color and the source reddens as it faints (MJD=55318). At the end of this period, the brightness increases again sharply and the slope in the CMD is comparable to the first period: The source becomes noticeably redder with higher luminosity. In the last 10 days of the monitoring the source again has a slope close to the ISM reddening law. Its luminosity continues to increase and the color becomes bluer until if has a similar color and luminosity as it did at the beginning of the monitoring. This closes the circle in the CMD. This is the only source in our sample where we observe multiple switches of the reddening slope over time. In this case variability similar to ISM reddening corresponds to more gradual changes in luminosity, while the other direction of the slope in the CMD is associated with faster luminosity changes.

One could thus suspect that in general sources that follow the ISM slope have a lower amplitude of variability than sources with a slope that is perpendicular in the CMD. However, comparing the amplitudes of all sources in the figure~\ref{fig:cmdslope} that follow the one or the other over the full time span of the observations we do not find significant differences in value or distribution of the variability amplitudes.

This study covers a larger number of stars, but uses only few photometric points to estimate $T_{eff}$.

At first sight, there seems to be a trend in the figure such that disks of cooler host stars are more variable. However, this is entirely caused by the different evolutionary stages. In very young objects (class~I or F) the SED is dominated by the other envelope or the accretion disk, thus the $T_{eff}$ in this case does not represent the effective temperature of the star. Section~\ref{sect:evolutionarytrendsofthevariability} already shows that earlier, more embedded systems have a strogner variability. Within a single class, e.g. class~II, there is no significant temperature dependence in figure~\ref{fig:teff}.
